## 1. O que é um vetor?

### 1.1 Intuição geométrica

No plano $\mathbb{R}^2$, um vetor é uma seta: tem direção e magnitude. Dois vetores podem ser somados (regra do paralelogramo) e escalados por um número real.

Mas essa intuição geométrica é um caso especial. O conceito de vetor é muito mais geral.

### 1.2 Definição algébrica

Um **vetor** é qualquer objeto que pode ser:
- **somado** com outro vetor do mesmo tipo, e
- **escalado** por um escalar (um número),

desde que essas operações satisfaçam os axiomas de espaço vetorial (associatividade, comutatividade, existência de neutro e inverso aditivo, distributividade da escala).

Em vez de apenas citar essas regras, podemos **testá-las explicitamente** em $\mathbb{R}^2$ para ver o que elas querem dizer na prática.

Exemplos que *são* vetores pela definição algébrica:
- Setas em $\mathbb{R}^2$ ou $\mathbb{R}^3$
- Tuplas de $n$ números reais: $(x_1, x_2, ..., x_n) \in \mathbb{R}^n$
- Funções contínuas $f$ — sim, funções são vetores
- Polinômios de grau $\leq n$
- Tuplas de $n$ números **complexos**: $(z_1, z_2, ..., z_n) \in \mathbb{C}^n$

```python
import numpy as np

def mostra_axioma(nome, lado_esquerdo, lado_direito):
    ok = np.allclose(lado_esquerdo, lado_direito)
    print(f"{nome}: {'OK' if ok else 'falhou'}")
    print('  lado esquerdo =', lado_esquerdo)
    print('  lado direito  =', lado_direito)
    print()

# Três vetores de R^2 para testar os axiomas
u = np.array([-4.0, 2.0])
v = np.array([2.0, 3.0])
w = np.array([1.0, -1.0])
a = 2.0
b = -0.5
zero = np.zeros(2)

print('Exemplo em R^2: vamos verificar alguns axiomas.')
print()

# Primeiro: as operações continuam dentro do espaço
print('Fechamento da soma:', v + w)
print('Fechamento por escalar:', 3 * v)
print()

mostra_axioma('Comutatividade da soma', v + w, w + v)
mostra_axioma('Associatividade da soma', (u + v) + w, u + (v + w))
mostra_axioma('Neutro aditivo', v + zero, v)
mostra_axioma('Inverso aditivo', v + (-v), zero)
mostra_axioma('Distributividade do escalar', a * (v + w), a * v + a * w)
mostra_axioma('Distributividade da soma de escalares', (a + b) * v, a * v + b * v)

# Qualquer combinação linear ainda é um vetor do mesmo espaço
combo = 0.6 * v + 0.4 * w
print('Combinação linear 0.6v + 0.4w =', combo)
print('Ela continua em R^2, então continua sendo um vetor desse espaço.')
```

## 2. Por que vetores precisam existir em um espaço?

Um vetor isolado não tem sentido matemático completo. As operações de soma e escala precisam de um **contexto** — um conjunto que:

1. É **fechado** sob as operações: somar dois vetores do conjunto produz outro vetor do mesmo conjunto.
2. Define quais são os **escalares válidos** (o corpo $\mathbb{F}$: $\mathbb{R}$, $\mathbb{C}$, etc.).
3. Garante a existência de um **elemento neutro** (vetor zero).

Esse contexto é o **espaço vetorial**. Sem ele, "somar dois vetores" é uma afirmação sem estrutura.

### Por que fechamento importa

Se você define "vetores" como pares $(x, y)$ com $x, y > 0$ (quadrante positivo), a soma $(1,2) + (-2,1)$ sai do conjunto. Ele **não é fechado** — e portanto nenhum teorema de álgebra linear se aplica a ele.

```python
import numpy as np

# R^2 é fechado
a = np.array([1.0, 2.0])
b = np.array([-2.0, 1.0])
print('a + b:', a + b)      # ainda em R^2
print('7.3a:', 7.3 * a)     # ainda em R^2
print()

# Quadrante positivo NÃO é espaço vetorial
x = np.array([1.0, 2.0])
y = np.array([-2.0, 1.0])
result = x + y
print(f'x + y = {result}')
print(f'Primeiro componente {result[0]:.0f} < 0: saiu do quadrante positivo.')
print('-> o quadrante positivo NÃO é espaço vetorial.')
```

## 3. O espaço complexo $\mathbb{C}^n$


### 3.1 De $\mathbb{R}^n$ para $\mathbb{C}^n$

Em $\mathbb{R}^n$ os escalares são números reais. Em $\mathbb{C}^n$ os escalares são números complexos $z = a + ib$, onde $a, b \in \mathbb{R}$ e $i^2 = -1$.

Ou seja, em $\mathbb{R}^n$ você só pode “multiplicar um vetor” por números da reta real (como 2, -3, 0.5). Já em $\mathbb{C}^n$, você passa a poder multiplicar por números que têm duas componentes: uma parte real ($a$) e uma parte imaginária ($b$). Isso é como sair de uma régua (1D) para um plano (2D) no conjunto dos números que você usa para escalar.

Todas as operações de $\mathbb{R}^n$ se generalizam, mas os escalares agora têm **fase** — o ângulo $\theta$ de $z = re^{i\theta}$ no plano complexo.

Ou seja, quando você multiplica um vetor por um número complexo, você não está apenas “esticando ou encolhendo” ele (como acontece com números reais). Você também pode **rotacionar** esse vetor no plano complexo. O termo $r$ controla o tamanho (magnitude), enquanto o ângulo $\theta$ controla essa rotação (fase).

Em termos práticos: em $\mathbb{R}^n$, escalar um vetor só muda o tamanho e o sinal. Em $\mathbb{C}^n$, escalar um vetor pode mudar o tamanho **e** introduzir uma rotação — e é exatamente essa “fase” que permite fenômenos como interferência na mecânica quântica.

> **Fase** vem do grego *phásis*, que significa “aparência” ou “manifestação”. Em contextos como ondas (luz, som), “fase” indica **em que ponto do ciclo** a onda está — por exemplo, se duas ondas estão alinhadas (em fase) ou desalinhadas (fora de fase).  
>  
> No caso de números complexos $z = re^{i\theta}$, o termo “fase” é usado porque o ângulo $\theta$ desempenha exatamente esse papel: ele indica **a posição no ciclo angular**.  
>  
> Ou seja, chamar $\theta$ de fase é herdar essa ideia física: não é só um ângulo geométrico, é um parâmetro que determina **como algo se alinha ou interfere com outro**.


### 3.2 Por que complexo importa


Números reais só têm sinal: $+$ ou $-$. Números complexos têm **fase contínua** $\theta \in [0, 2\pi)$.

Ou seja, com números reais você só consegue duas “direções”: para frente (positivo) ou para trás (negativo). Já com números complexos, você tem um **círculo inteiro de direções possíveis**. Cada número complexo pode ser visto como uma seta no plano, com um tamanho (magnitude) e uma direção (fase).

> **Cuidado com a intuição de “ângulo”**: ângulo não nasce com números complexos — ele já existe em $\mathbb{R}^2$ e na trigonometria (círculo, seno, cosseno, etc.).  
>  
> A diferença aqui não é geométrica, é estrutural. Em $\mathbb{R}^n$, o ângulo é sempre algo que você define **entre dois vetores**. Já os escalares (números reais) não carregam direção — eles só mudam o tamanho (ou o sinal) de um vetor.  
>  
> Em $\mathbb{C}$, cada número $z = re^{i\theta}$ já vem com uma magnitude $r$ e um ângulo $\theta$. Esse ângulo — a fase — é medido em relação a um eixo de referência (tipicamente o eixo real), mas o ponto importante é que ele está **embutido no próprio número**.  
>  
> Ou seja: o ângulo continua sendo, em essência, um ângulo no plano. A diferença é que, no caso complexo, esse ângulo passa a ser um atributo do escalar que você usa nas operações, e não algo que você calcula externamente entre vetores.  

Você pode, sim, representar fenômenos usando apenas números reais (seno, cosseno, etc.). Então a pergunta correta não é “dá para fazer com reais?”, porque dá. A pergunta correta é: **o que fica difícil ou artificial usando apenas reais?** Para responder isso, precisamos olhar para o tipo de fenômeno que queremos modelar. Vou propor um: `onda`.

Na física, ondas são sempre alguma forma de energia ou grandeza física que oscila no tempo (e/ou no espaço). Exemplos:

- **vibração de uma corda** → a posição (deslocamento) de cada ponto da corda oscila, armazenando energia cinética e potencial elástica  
- **onda sonora** → a pressão do ar (e a densidade do meio) oscila, transportando energia mecânica  
- **campo eletromagnético (luz)** → os campos elétrico e magnético oscilam no espaço e no tempo, transportando energia eletromagnética  
- **função de onda (mecânica quântica)** → a amplitude complexa oscila; o que ela codifica são probabilidades de medição, e sua fase governa interferência

Matematicamente, uma onda pode ser escrita como:

$$
f(t) = A \cos(\omega t + \theta)
$$

- uma amplitude $A$ (o “tamanho” da oscilação),
- uma frequência $\omega$ (quão rápido oscila),
- e uma fase $\theta$ (em que ponto do ciclo ela está).

Agora vem o problema com números reais.

Se você quiser manipular essa onda (somar, deslocar fase, evoluir no tempo), a forma $\cos(\omega t + \theta)$ não é muito conveniente. Você acaba precisando reescrever tudo usando identidades trigonométricas, e isso rapidamente fica complicado. Equivalentemente, você pode escrever a mesma onda como combinação de seno e cosseno:

- seno → $\sin(\theta)$
- cosseno → $\cos(\theta)$

Ou seja, uma única “onda” já exige dois números. Com números complexos, essas duas informações são unificadas:
$$
e^{i\theta} = \cos(\theta) + i \sin(\theta)
$$

> Essa igualdade é conhecida como **identidade de Euler**. Ela conecta exponenciais complexas com funções trigonométricas, mostrando que seno e cosseno podem ser vistos como partes de uma única estrutura mais fundamental: uma rotação no plano complexo.

Ou seja, um único número carrega simultaneamente:
- magnitude (amplitude)
- fase (posição no ciclo)

Agora a mudança importante: com essa representação, **evoluir no tempo vira uma multiplicação simples**. Por exemplo, uma onda que evolui no tempo:
$$
e^{i(\omega t + \theta)} = e^{i\theta} \cdot e^{i\omega t}
$$

Ou seja, “andar no tempo” corresponde a **multiplicar por um número complexo**. Isso não é só conveniência — isso revela a estrutura correta do problema:
- oscilações são rotações,
- e rotações são naturalmente descritas no plano complexo.

Em mecânica quântica, isso é essencial. A evolução do sistema precisa ser:
- linear,
- reversível (unitária),

e essas evoluções são exatamente **rotações no espaço de estados complexo**. Com números reais, você até consegue representar tudo isso, mas:
- precisa separar seno e cosseno (duas variáveis),
- perde a representação direta da fase como um único parâmetro,
- e a evolução deixa de ser uma operação simples (vira manipulação trigonométrica).

Ou seja: não é que números reais não funcionam — é que **números complexos capturam diretamente a natureza oscilatória das ondas como rotações**, de forma mais simples e estruturalmente correta.

E como a mecânica quântica descreve sistemas que se comportam como ondas, essa escolha deixa de ser estética e passa a ser natural.

```python
import numpy as np

print("=== Real vs Complexo: representando uma onda ===\n")

theta = np.pi / 3  # fase qualquer

# Representação real (duas componentes)
real_cos = np.cos(theta)
real_sin = np.sin(theta)

print("Representação real:")
print("cos(θ) =", real_cos)
print("sin(θ) =", real_sin)

# Representação complexa (uma só)
z = np.exp(1j * theta)

print("\nRepresentação complexa:")
print("e^{iθ} =", z)

print("\nConclusão:")
print("Reais → precisa de 2 números (cos, sin)")
print("Complexo → 1 número carrega ambos (magnitude + fase)")
```

## 4. O Espaço de Hilbert

### 4.1 O que adiciona ao espaço complexo

Um **espaço de Hilbert** $\mathcal{H}$ é um espaço vetorial sobre $\mathbb{C}$ equipado com um **produto interno** $\langle \cdot, \cdot \rangle$ que satisfaz:

1. **Linear no segundo argumento:** $\langle u, \alpha v + \beta w \rangle = \alpha\langle u,v\rangle + \beta\langle u,w\rangle$
2. **Conjugado-simétrico:** $\langle u, v \rangle = \overline{\langle v, u \rangle}$
3. **Positivo definido:** $\langle v, v \rangle \geq 0$, com igualdade só se $v = 0$
4. **Completo:** toda sequência de Cauchy converge dentro do espaço

A completude distingue Hilbert de um espaço pré-Hilbert. Para **dimensão finita** — o caso de $n$ qubits, com $\mathcal{H} = \mathbb{C}^{2^n}$ — todo espaço com produto interno é automaticamente completo. A distinção importa em dimensão infinita (mecânica quântica contínua, análise funcional).

**Para computação quântica com $n$ qubits:** $\mathcal{H} = \mathbb{C}^{2^n}$ com produto interno padrão. É essencialmente $\mathbb{C}^{2^n}$ com a estrutura certa.

### 4.2 O que o produto interno compra

| Conceito | Definição via produto interno | Uso quântico |
|---|---|---|
| Norma | $\|v\| = \sqrt{\langle v, v \rangle}$ | Conservação de probabilidade |
| Ortogonalidade | $\langle u, v \rangle = 0$ | Estados distinguíveis por medição |
| Projeção | $\frac{\langle u, v \rangle}{\langle u, u \rangle} u$ | O que medição faz ao estado |

```python
import numpy as np

# H = C^2 (espaço de 1 qubit)
ket_0 = np.array([1.0, 0.0], dtype=complex)
ket_1 = np.array([0.0, 1.0], dtype=complex)

print('Ortogonalidade da base computacional:')
print('  ⟨0|0⟩ =', np.dot(ket_0.conj(), ket_0))  # 1
print('  ⟨1|1⟩ =', np.dot(ket_1.conj(), ket_1))  # 1
print('  ⟨0|1⟩ =', np.dot(ket_0.conj(), ket_1))  # 0 <- ortogonais
print()

# Projeção = o que medição faz
psi = np.array([0.6, 0.8], dtype=complex)
print(f'|ψ⟩ = {psi},  norma = {np.linalg.norm(psi):.6f}')
print()

amp_0 = np.dot(ket_0.conj(), psi)   # ⟨0|ψ⟩
amp_1 = np.dot(ket_1.conj(), psi)   # ⟨1|ψ⟩
prob_0 = abs(amp_0)**2
prob_1 = abs(amp_1)**2

print(f'⟨0|ψ⟩ = {amp_0}  ->  P(|0⟩) = {prob_0:.4f}')
print(f'⟨1|ψ⟩ = {amp_1}  ->  P(|1⟩) = {prob_1:.4f}')
print(f'P(|0⟩) + P(|1⟩) = {prob_0 + prob_1:.6f}  (= 1: conservação de probabilidade)')
```

## 5. O vetor de estado e por que ele é usado na mecânica quântica

### 5.1 O que é $|\psi\rangle$

O **vetor de estado** $|\psi\rangle \in \mathcal{H}$ é um vetor normalizado ($\||\psi\rangle\| = 1$) que contém toda a informação física acessível sobre o sistema.

Para 1 qubit: $|\psi\rangle = \alpha|0\rangle + \beta|1\rangle$ com $|\alpha|^2 + |\beta|^2 = 1$.

A notação $|\cdot\rangle$ (ket) é convenção de Dirac. O bra é $\langle\psi| = |\psi\rangle^\dagger$, de modo que o produto interno é $\langle\phi|\psi\rangle$.

### 5.2 Por que um vetor — e não outra coisa?

**1. Superposição é linear.** Se $|\psi_1\rangle$ e $|\psi_2\rangle$ são estados válidos, $\alpha|\psi_1\rangle + \beta|\psi_2\rangle$ (normalizado) também é. Isso é o princípio da superposição — e é exatamente o que espaços vetoriais garantem.

**2. Probabilidades vêm do produto interno.** $P(|\phi\rangle) = |\langle\phi|\psi\rangle|^2$. A norma garantida pelo espaço de Hilbert faz as probabilidades somarem 1.

**3. Fase é física.** A fase relativa entre $\alpha$ e $\beta$ afeta resultados de medição. $\mathbb{C}^n$ preserva essa fase. $\mathbb{R}^n$ não seria suficiente.

**4. Operadores são transformações lineares.** Portas quânticas, evolução temporal, observáveis — todos são operadores lineares sobre $\mathcal{H}$. A estrutura vetorial é o que permite escrever $U|\psi\rangle$.

### 5.3 O que o vetor de estado não é

| Confusão comum | Por que está errado |
|---|---|
| "É uma distribuição de probabilidades" | As amplitudes $\alpha, \beta$ são complexas. A fase importa e não aparece nas probabilidades $|\alpha|^2, |\beta|^2$. |
| "É o sistema físico" | É uma representação matemática do estado de conhecimento máximo sobre o sistema. |
| "Fase global importa" | $e^{i\theta}|\psi\rangle$ é fisicamente idêntico a $|\psi\rangle$. Só fase **relativa** é observável. |
| "É um vetor real" | Amplitudes são complexas. Interferência requer fase contínua — impossível em $\mathbb{R}^n$. |

```python
import numpy as np

# Fase global vs. fase relativa
psi = np.array([1, 1], dtype=complex) / np.sqrt(2)  # |+>

# Fase global: multiplica o estado inteiro
psi_global = np.exp(1j * np.pi/3) * psi

# Fase relativa: multiplica só |1>
psi_relat = np.array([psi[0], psi[1] * np.exp(1j * np.pi/3)])

ket_0    = np.array([1.0, 0.0], dtype=complex)
ket_plus = np.array([1.0, 1.0], dtype=complex) / np.sqrt(2)

def p(state, basis_vec):
    return abs(np.dot(basis_vec.conj(), state))**2

print('Medição na base Z — P(|0⟩):')
print(f'  |ψ⟩ original:      {p(psi, ket_0):.4f}')
print(f'  fase global:       {p(psi_global, ket_0):.4f}  <- idêntico')
print(f'  fase relativa:     {p(psi_relat, ket_0):.4f}  <- idêntico')
print()
print('Medição na base X — P(|+⟩):')
print(f'  |ψ⟩ original:      {p(psi, ket_plus):.4f}')
print(f'  fase global:       {p(psi_global, ket_plus):.4f}  <- idêntico')
print(f'  fase relativa:     {p(psi_relat, ket_plus):.4f}  <- diferente!')
print()
print('Conclusão: fase global é inobservável. Fase relativa é física.')
```

## Por que isso importa para algoritmos quânticos?

**1. Superposição é vetorial, não probabilística.** $|\psi\rangle = \alpha|0\rangle + \beta|1\rangle$ não é "50% em 0 e 50% em 1". É um vetor com amplitudes complexas. A diferença aparece quando dois estados interagem: as amplitudes interferem, as probabilidades não.

**2. Portas quânticas são transformações lineares.** $U|\psi\rangle$ age sobre cada componente da superposição simultaneamente. O que parece "paralelismo quântico" é linearidade de operadores em $\mathcal{H}$.

**3. Medição é projeção.** $P(|\phi\rangle) = |\langle\phi|\psi\rangle|^2$ é o quadrado do produto interno. A estrutura do espaço de Hilbert garante que a soma sobre todos os resultados é 1 — conservação de probabilidade.

**4. Phase kickback exige $\mathbb{C}$.** O mecanismo central de C-03 move uma fase complexa entre registradores. Sem $\mathbb{C}^n$, não há fase para mover.

**5. A dimensão cresce exponencialmente.** Para $n$ qubits, $\mathcal{H} = \mathbb{C}^{2^n}$. Um estado de 50 qubits tem $2^{50} \approx 10^{15}$ amplitudes complexas — impossível de armazenar classicamente. Esse crescimento é simultaneamente o poder e o limite da computação quântica.
