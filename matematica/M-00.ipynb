## 1. O que é um vetor?

A noção de vetor surgiu inicialmente na física, especialmente nos séculos XVII–XIX, para descrever grandezas que possuem **magnitude e direção**, como força, velocidade e aceleração. No início, vetores eram entendidos de forma puramente geométrica — como setas no espaço. Ao longo do século XIX, com o desenvolvimento da álgebra linear, matemáticos como Hermann Grassmann e Giuseppe Peano perceberam que o conceito podia ser abstraído: o importante não era a “seta” em si, mas as **operações** que esses objetos obedeciam (soma e multiplicação por escalares). Isso levou à definição moderna de espaço vetorial, onde vetores passaram a ser qualquer objeto que satisfaz certos axiomas — não apenas pontos no espaço, mas também funções, sequências e polinômios. Essa abstração foi essencial para unificar diferentes áreas da matemática e, mais tarde, fornecer a linguagem adequada para teorias físicas como a mecânica quântica.

### 1.1 Intuição geométrica

No plano $\mathbb{R}^2$, um vetor é uma seta: tem direção e magnitude. Dois vetores podem ser somados (regra do paralelogramo) e escalados por um número real.

Mas essa intuição geométrica é um caso especial. O conceito de vetor é muito mais geral.

### 1.2 Definição algébrica

Um **vetor** é qualquer objeto que pode ser:
- **somado** com outro vetor do mesmo tipo, e
- **escalado** por um escalar (um número),

desde que essas operações satisfaçam os axiomas de espaço vetorial (associatividade, comutatividade, existência de neutro e inverso aditivo, distributividade da escala).

Em vez de apenas citar essas regras, podemos **testá-las explicitamente** em $\mathbb{R}^2$ para ver o que elas querem dizer na prática.

Exemplos que *são* vetores pela definição algébrica:
- Setas em $\mathbb{R}^2$ ou $\mathbb{R}^3$
- Tuplas de $n$ números reais: $(x_1, x_2, ..., x_n) \in \mathbb{R}^n$
- Funções contínuas $f$ — sim, funções são vetores
- Polinômios de grau $\leq n$
- Tuplas de $n$ números **complexos**: $(z_1, z_2, ..., z_n) \in \mathbb{C}^n$

## 2. Por que vetores precisam existir em um espaço?

A ideia de “espaço” surge da necessidade de organizar e dar sentido aos vetores. No início, no contexto da geometria euclidiana (Antiguidade até o século XVII), espaço significava simplesmente o ambiente físico onde pontos e setas (vetores) vivem — como o plano ou o espaço tridimensional. Com o avanço da matemática nos séculos XVIII e XIX, ficou claro que era possível trabalhar com objetos mais abstratos (como sequências, funções e polinômios) que ainda obedeciam às mesmas regras de combinação dos vetores. Para tratar esses objetos de forma consistente, foi necessário definir explicitamente o “lugar” onde eles vivem: um conjunto com estrutura bem definida, onde operações como soma e multiplicação por escalares fazem sentido e obedecem regras específicas. Esse “lugar estruturado” é o que chamamos de espaço. Ou seja, o espaço não é apenas um cenário geométrico, mas a estrutura que garante que os vetores possam ser manipulados de forma coerente — permitindo generalizar a geometria para contextos muito além do físico.

Um vetor isolado não tem sentido matemático completo. As operações de soma e escala precisam de um **contexto** — um conjunto que:

1. É **fechado** sob as operações: somar dois vetores do conjunto produz outro vetor do mesmo conjunto.
2. Define quais são os **escalares válidos** (o corpo $\mathbb{F}$: $\mathbb{R}$, $\mathbb{C}$, etc.).
3. Garante a existência de um **elemento neutro** (vetor zero).

Esse contexto é o **espaço vetorial**. Sem ele, "somar dois vetores" é uma afirmação sem estrutura.

### Por que fechamento importa

Se você define "vetores" como pares $(x, y)$ com $x, y > 0$ (quadrante positivo), a soma $(1,2) + (-2,1)$ sai do conjunto. Ele **não é fechado** — e portanto nenhum teorema de álgebra linear se aplica a ele.

## 3. O espaço complexo $\mathbb{C}^n$


A ideia de trabalhar com o espaço complexo $\mathbb{C}^n$ não surgiu diretamente da física, mas de uma necessidade matemática mais antiga: lidar com equações que não tinham solução no conjunto dos números reais. No século XVI, matemáticos italianos como Gerolamo Cardano e Rafael Bombelli começaram a manipular raízes de números negativos ao estudar equações cúbicas — mesmo sem entender completamente o que esses “números imaginários” eram. Ao longo dos séculos XVIII e XIX, com contribuições de Euler e Gauss, os números complexos ganharam uma interpretação geométrica sólida como pontos no plano. A partir daí, tornou-se natural considerar vetores cujas coordenadas são complexas, formando $\mathbb{C}^n$. Esse espaço não é apenas uma extensão técnica de $\mathbb{R}^n$: ele carrega uma estrutura mais rica, onde multiplicação por escalares envolve não só escala, mas também rotação (fase). Essa propriedade acabou sendo essencial para descrever fenômenos oscilatórios e, mais tarde, tornou $\mathbb{C}^n$ o ambiente natural para a formulação da mecânica quântica.

### 3.1 De $\mathbb{R}^n$ para $\mathbb{C}^n$

Em $\mathbb{R}^n$ os escalares são números reais. Em $\mathbb{C}^n$ os escalares são números complexos $z = a + ib$, onde $a, b \in \mathbb{R}$ e $i^2 = -1$.

Ou seja, em $\mathbb{R}^n$ você só pode "multiplicar um vetor" por números da reta real (como 2, -3, 0.5). Já em $\mathbb{C}^n$, você passa a poder multiplicar por números que têm duas componentes: uma parte real ($a$) e uma parte imaginária ($b$). Isso é como sair de uma régua (1D) para um plano (2D) no conjunto dos números que você usa para escalar.

Todas as operações de $\mathbb{R}^n$ se generalizam, mas os escalares agora têm [**fase**](../M-00.ipynb#C3:L1)  — o ângulo $\theta$ de $z = re^{i\theta}$ no plano complexo.

Ou seja, quando você multiplica um vetor por um número complexo, você não está apenas "esticando ou encolhendo" ele (como acontece com números reais). Você também pode **rotacionar** esse vetor no plano complexo. O termo $r$ controla o tamanho (magnitude), enquanto o ângulo $\theta$ controla essa rotação (fase).

Em termos práticos: em $\mathbb{R}^n$, escalar um vetor só muda o tamanho e o sinal. Em $\mathbb{C}^n$, escalar um vetor pode mudar o tamanho **e** introduzir uma rotação — e é exatamente essa "fase" que permite fenômenos como interferência na mecânica quântica.

### 3.2 Por que complexo importa


Números reais só têm sinal: $+$ ou $-$. Números complexos têm **fase contínua** $\theta \in [0, 2\pi)$.

Ou seja, com números reais você só consegue duas “direções”: para frente (positivo) ou para trás (negativo). Já com números complexos, você tem um **círculo inteiro de direções possíveis**. Cada número complexo pode ser visto como uma seta no plano, com um tamanho (magnitude) e uma direção (fase).

> **Cuidado com a intuição de “ângulo”**: ângulo não nasce com números complexos — ele já existe em $\mathbb{R}^2$ e na trigonometria (círculo, seno, cosseno, etc.).  
>  
> A diferença aqui não é geométrica, é estrutural. Em $\mathbb{R}^n$, o ângulo é sempre algo que você define **entre dois vetores**. Já os escalares (números reais) não carregam direção — eles só mudam o tamanho (ou o sinal) de um vetor.  
>  
> Em $\mathbb{C}$, cada número $z = re^{i\theta}$ já vem com uma magnitude $r$ e um ângulo $\theta$. Esse ângulo — a fase — é medido em relação a um eixo de referência (tipicamente o eixo real), mas o ponto importante é que ele está **embutido no próprio número**.  
>  
> Ou seja: o ângulo continua sendo, em essência, um ângulo no plano. A diferença é que, no caso complexo, esse ângulo passa a ser um atributo do escalar que você usa nas operações, e não algo que você calcula externamente entre vetores.  

Você pode, sim, representar fenômenos usando apenas números reais (seno, cosseno, etc.). Então a pergunta correta não é “dá para fazer com reais?”, porque dá. A pergunta correta é: **o que fica difícil ou artificial usando apenas reais?** Para responder isso, precisamos olhar para o tipo de fenômeno que queremos modelar. Vou propor um: `onda`.

Na física, ondas são sempre alguma forma de energia ou grandeza física que oscila no tempo (e/ou no espaço). Exemplos:

- **vibração de uma corda** → a posição (deslocamento) de cada ponto da corda oscila, armazenando energia cinética e potencial elástica  
- **onda sonora** → a pressão do ar (e a densidade do meio) oscila, transportando energia mecânica  
- **campo eletromagnético (luz)** → os campos elétrico e magnético oscilam no espaço e no tempo, transportando energia eletromagnética  
- **função de onda (mecânica quântica)** → a amplitude complexa oscila; o que ela codifica são probabilidades de medição, e sua fase governa interferência

Matematicamente, uma onda pode ser escrita como:

$$
f(t) = A \cos(\omega t + \theta)
$$

- uma amplitude $A$ (o “tamanho” da oscilação),
- uma frequência $\omega$ (quão rápido oscila),
- e uma fase $\theta$ (em que ponto do ciclo ela está).

Agora vem o problema com números reais.

Se você quiser manipular essa onda (somar, deslocar fase, evoluir no tempo), a forma $\cos(\omega t + \theta)$ não é muito conveniente. Você acaba precisando reescrever tudo usando identidades trigonométricas, e isso rapidamente fica complicado. Equivalentemente, você pode escrever a mesma onda como combinação de seno e cosseno:

- seno → $\sin(\theta)$
- cosseno → $\cos(\theta)$

Ou seja, uma única “onda” já exige dois números. Com números complexos, essas duas informações são unificadas:
$$
e^{i\theta} = \cos(\theta) + i \sin(\theta)
$$

> Essa igualdade é conhecida como **identidade de Euler**. Ela conecta exponenciais complexas com funções trigonométricas, mostrando que seno e cosseno podem ser vistos como partes de uma única estrutura mais fundamental: uma rotação no plano complexo.

A razão disso funcionar está na própria definição dos números complexos: ao introduzir o número imaginário $i$, tal que $i^2 = -1$, nós essencialmente construímos um plano onde multiplicar por $i$ corresponde a uma rotação de $90^\circ$. Ou seja, o “eixo imaginário” não é apenas um segundo eixo arbitrário — ele foi definido de forma que a operação fundamental (multiplicação) já incorpore rotação. Quando você combina isso com exponenciais, $e^{i\theta}$ passa a representar rotações contínuas nesse plano. Em outras palavras: os números complexos conseguem unificar seno e cosseno porque sua própria álgebra foi construída para que multiplicar números signifique girar vetores — e ondas são exatamente isso, rotações contínuas.

Ou seja: não é que números reais não funcionam — é que **números complexos capturam diretamente a natureza oscilatória das ondas como rotações**, de forma mais simples e estruturalmente correta.

## 4. O Espaço de Hilbert

O conceito de **espaço de Hilbert** leva o nome de David Hilbert (1862–1943), um matemático alemão que, no início do século XX (por volta de 1900–1920), estava tentando formalizar problemas envolvendo **equações integrais** — um tipo de equação que aparece naturalmente na física, especialmente em teoria do potencial e mecânica. A dificuldade era que essas equações não podiam ser tratadas apenas com vetores finitos como em $\mathbb{R}^n$, mas exigiam trabalhar com objetos como **funções**, que formam espaços de dimensão infinita.

Hilbert percebeu que era possível estender as ideias de geometria (comprimento, ângulo, ortogonalidade) para esses espaços mais gerais, desde que existisse um produto interno bem definido. Esse esforço de generalização levou à noção de espaço de Hilbert: um espaço vetorial **sobre $\mathbb{C}$** (ou $\mathbb{R}$, em versões mais gerais), equipado com produto interno e propriedades de completude, capaz de tratar funções como “vetores”. Décadas depois, essa estrutura se tornou a linguagem natural da mecânica quântica, onde estados físicos são modelados exatamente como vetores nesse tipo de espaço — especificamente, em espaços de Hilbert complexos.

### 4.1 O que adiciona ao espaço complexo

Um **espaço de Hilbert** $\mathcal{H}$ é um espaço vetorial sobre $\mathbb{C}$ equipado com um **produto interno** $\langle \cdot, \cdot \rangle$ que satisfaz:

1. **Linear no segundo argumento:** $\langle u, \alpha v + \beta w \rangle = \alpha\langle u,v\rangle + \beta\langle u,w\rangle$
2. **Conjugado-simétrico:** $\langle u, v \rangle = \overline{\langle v, u \rangle}$
3. **Positivo definido:** $\langle v, v \rangle \geq 0$, com igualdade só se $v = 0$
4. **Completo:** toda sequência de Cauchy converge dentro do espaço

> Uma **sequência de Cauchy** é uma sequência onde os termos vão ficando cada vez mais próximos entre si à medida que avançamos.  Ou seja, depois de certo ponto, todos os elementos da sequência estão “quase iguais”, mesmo que você não saiba ainda qual é o valor final.  Dizer que o espaço é completo significa: se uma sequência está se estabilizando assim, então o limite dela **existe dentro do próprio espaço**.

### 4.2 O que o produto interno compra

Por que o nome **produto interno**?  

Porque ele pega dois vetores do próprio espaço e devolve um número que mede como eles se relacionam — sem sair do espaço onde eles vivem. Diferente de outras operações que podem envolver objetos externos, o produto interno é uma forma de “comparar vetores por dentro”: ele mede alinhamento, magnitude e sobreposição usando apenas a estrutura do próprio espaço.  

Em termos práticos: $\langle u, v \rangle$ responde “o quanto $u$ aponta na direção de $v$”. Tudo que vem depois — norma, ângulo, projeção — nasce dessa única operação. Com efeito, a “geometria” (comprimento, ângulo, projeção) já existia de forma intuitiva. O que acontece aqui é que ela ganha um formalismo: tudo passa a ser definido a partir do produto interno. Ou seja, em vez de assumir noções geométricas, você as reconstrói a partir de uma única operação algébrica.

| Conceito | Definição via produto interno | Intuição |
|---|---|---|
| Norma | $\|v\| = \sqrt{\langle v, v \rangle}$ | “Tamanho” do vetor |
| Ortogonalidade | $\langle u, v \rangle = 0$ | Vetores perpendiculares (não se influenciam) |
| Projeção | $\frac{\langle u, v \rangle}{\langle u, u \rangle} u$ | Quanto de $v$ está na direção de $u$ |

## 5. O vetor de estado

A noção de **vetor de estado** surgiu no contexto da formulação matemática da mecânica quântica na década de 1920, principalmente com o trabalho de John von Neumann. A teoria quântica já havia sido desenvolvida por físicos como Schrödinger (equação de onda) e Heisenberg (mecânica matricial), mas faltava uma estrutura matemática unificada. Von Neumann percebeu que ambos os formalismos podiam ser descritos de forma consistente usando vetores em espaços de Hilbert: o estado de um sistema físico passa a ser representado por um vetor, e observáveis por operadores nesses vetores. Ou seja, o “vetor de estado” não foi inventado do zero, mas emerge da aplicação da teoria de espaços vetoriais e de Hilbert para organizar e formalizar a mecânica quântica.

### 5.1 O que é $|\psi\rangle$

O **vetor de estado** $|\psi\rangle \in \mathcal{H}$ é um objeto matemático que usamos para representar **o estado de um sistema físico**. Que sistema?
- posição e momento de uma partícula (mecânica quântica contínua)
- polarização de um fóton (horizontal / vertical)

Ou seja: qualquer sistema físico que pode estar em diferentes configurações possíveis — o vetor de estado é a forma de codificar isso. Agora, por que ele é normalizado? Porque o que realmente importa não são os coeficientes em si, mas os **módulos ao quadrado** desses coeficientes — eles representam probabilidades. Então exigir que $\||\psi\rangle\| = 1$ é garantir que a soma dessas probabilidades dá 1:
$$
\langle \psi | \psi \rangle = 1
$$

Ou seja: é uma restrição de consistência — “toda a probabilidade está dentro do sistema”. E o que exatamente ele armazena? Ele não armazena “valores físicos” diretamente. Ele armazena algo mais fundamental: **como o sistema responde quando você faz perguntas sobre ele**. Mais concretamente:
- cada direção (vetor base) representa uma possível configuração
- os coeficientes dizem “o quanto” o estado está alinhado com cada uma dessas direções

Então o vetor de estado é, essencialmente:
> uma codificação de todas as possíveis respostas do sistema a medições, organizada de forma linear

Para o caso específico de 1 qubit:
$$
|\psi\rangle = \alpha|0\rangle + \beta|1\rangle, \quad |\alpha|^2 + |\beta|^2 = 1
$$

A notação $|\cdot\rangle$ (ket) é convenção de Dirac. O bra é $\langle\psi| = |\psi\rangle^\dagger$, e o produto interno $\langle\phi|\psi\rangle$ mede o quanto dois estados “se sobrepõem”.

### 5.2 Por que um vetor — e não outra coisa?

**1. Superposição é linear.** Se $|\psi_1\rangle$ e $|\psi_2\rangle$ são estados válidos, $\alpha|\psi_1\rangle + \beta|\psi_2\rangle$ (normalizado) também é. Isso é o princípio da superposição — e é exatamente o que espaços vetoriais garantem.

**2. Probabilidades vêm do produto interno.** $P(|\phi\rangle) = |\langle\phi|\psi\rangle|^2$. A norma garantida pelo espaço de Hilbert faz as probabilidades somarem 1.

**3. Fase é física.** A fase relativa entre $\alpha$ e $\beta$ afeta resultados de medição. $\mathbb{C}^n$ preserva essa fase. $\mathbb{R}^n$ não seria suficiente.

**4. Operadores são transformações lineares.** Portas quânticas, evolução temporal, observáveis — todos são operadores lineares sobre $\mathcal{H}$. A estrutura vetorial é o que permite escrever $U|\psi\rangle$.

### 5.3 O que o vetor de estado não é

- **"É uma distribuição de probabilidades" (não é)**  
  As amplitudes $\alpha$ e $\beta$ são complexas. As probabilidades vêm de $|\alpha|^2$ e $|\beta|^2$, mas a fase (que não aparece nelas) influencia a interferência.

- **"É o sistema físico" (não é)**  
  É uma representação matemática do estado do sistema — não o sistema em si.

- **"Fase global importa" (não importa)**  
  Multiplicar o estado por uma fase global $e^{i\theta}$ não muda nenhuma probabilidade:  
  $|\langle \phi | e^{i\theta}\psi \rangle|^2 = |\langle \phi | \psi \rangle|^2$.  
  Logo, $e^{i\theta}\lvert\psi\rangle$ e $\lvert\psi\rangle$ são fisicamente indistinguíveis.  
  Apenas diferenças de fase entre componentes (fase relativa) afetam interferência.

- **"É um vetor real" (não é)**  
  As amplitudes são complexas. A fase contínua é essencial para descrever interferência — algo que $\mathbb{R}^n$ não captura naturalmente.

Um vetor é qualquer objeto que pode ser somado com outro de seu tipo e escalado por um número, desde que essas operações obedeçam um conjunto fixo de regras.

Para que essas operações façam sentido, os vetores precisam habitar um espaço vetorial: um conjunto fechado sob soma e escala, com elemento neutro e escalares bem definidos.

Quando os escalares são números complexos, cada vetor carrega não só magnitude mas fase — o ângulo no plano complexo.

A fase é o que torna possível a interferência: duas amplitudes de mesma magnitude podem se cancelar completamente se apontarem em direções opostas no plano complexo.

Para medir distância e ângulo entre vetores complexos é preciso um produto interno; em $\mathbb{C}^n$ ele conjuga o primeiro vetor, garantindo que $\langle v, v \rangle$ seja sempre real e não-negativo.

O espaço de Hilbert é um espaço vetorial complexo com produto interno que satisfaz quatro propriedades: linearidade, simetria conjugada, positividade e completude.

O produto interno define norma, ortogonalidade e projeção — os três instrumentos que a mecânica quântica usa para falar de probabilidade, estados distinguíveis e medição.

O vetor de estado $|\psi\rangle$ é um vetor normalizado nesse espaço: representa o estado de conhecimento máximo sobre o sistema, não o sistema físico em si.

Fase global $e^{i\theta}|\psi\rangle$ é inobservável; fase relativa entre componentes é física e afeta resultados de medição — e é exatamente ela que os algoritmos quânticos manipulam.
