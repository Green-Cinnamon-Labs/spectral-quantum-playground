## Introdução

A álgebra linear não foi planejada — ela emergiu da necessidade de resolver equações, depois de descrever transformações, depois de falar sobre espaços de qualquer dimensão, e por fim de lidar com objetos que se transformam de maneiras mais ricas que vetores simples.

**Transformação** é um termo bastante presente na álgebra, porque via de regra ela se dedica a entender como pegar um objeto matemático e converter em outro. Isso é extremamente útil em muitos casos.

Este notebook percorre essa trajetória do início: da álgebra como arte de manipular símbolos (al-Khwārizmī, séc. IX) até o cálculo tensorial (Ricci, séc. XIX). O fio condutor é sempre o mesmo: cada abstração surgiu porque a anterior não era suficiente para descrever o que os matemáticos estavam tentando entender.

## 1. O surgimento da álgebra: al-Khwārizmī e a arte de restaurar equações

Em 830 d.C., o matemático persa Muhammad ibn Mūsā al-Khwārizmī escreve em Bagdá um tratado sobre como resolver equações do segundo grau. O livro não usa símbolos — usa palavras. "Uma quadrado mais dez raízes é igual a trinta e nove."

O passo revolucionário não é o resultado — é o método. Al-Khwārizmī percebe que certas operações se repetem em todo tipo de equação: mover um termo para o outro lado (`al-jabr`, restauração) e simplificar termos semelhantes (`al-muqābala`, equilíbrio). Essas operações independem do problema específico — funcionam para qualquer equação da mesma forma.

Isso é o embrião do pensamento algébrico: **a estrutura da manipulação importa mais que os números específicos**.

Nos séculos seguintes, matemáticos europeus (Fibonacci, Cardano, Viète) traduzem e estendem esse trabalho, introduzindo progressivamente a notação simbólica. Viète (séc. XVI) é o primeiro a usar letras sistematicamente para representar quantidades desconhecidas — *incógnitas*. Descartes (séc. XVII) codifica a convenção que usamos hoje: letras do fim do alfabeto ($x, y, z$) para incógnitas, do início ($a, b, c$) para parâmetros.

Mas essa álgebra ainda é sobre números — sobre encontrar valores que satisfaçam equações. A grande ruptura vem no século XIX.

## 2. A abstração: álgebra sobre estruturas, não números

Em 1832, Évariste Galois — com 20 anos, na véspera de um duelo em que morreria — escreve uma carta descrevendo a teoria que hoje leva seu nome. Ele estava investigando por que equações de grau 5 não têm solução por radicais e percebeu que a resposta dependia da estrutura de `simetria` das raízes, não das raízes em si.

Para capturar essa estrutura, Galois trabalha com `grupos`: conjuntos com uma operação que satisfaz quatro axiomas — fechamento, associatividade, elemento neutro e inverso. A pergunta não é mais "qual é o valor de $x$?" mas "qual é a estrutura do conjunto de simetrias deste objeto?"

Essa mudança de perspectiva é profunda. A álgebra deixa de ser sobre encontrar números e passa a ser sobre estudar estruturas. Na sequência:

- **Grupos** (Galois, 1832): uma operação, quatro axiomas
- **Anéis** (Dedekind, 1870s): dois operações (adição + multiplicação), sem inverso multiplicativo garantido
- **Corpos** (Weber, 1893): dois operações com todos os inversos — $\mathbb{Q}$, $\mathbb{R}$, $\mathbb{C}$ são corpos
- **Espaços vetoriais** (Peano, 1888): um conjunto com adição e multiplicação por escalares de um corpo

Cada estrutura é um nível de abstração acima da anterior. E cada uma surge porque a anterior não era suficiente para o problema em questão.

<div style="border-left: 2px solid #f59e0b; padding: 0.5em 1em 0.5em 1.1em; margin: 0.8em 0 1.2em 0; background: rgba(245, 158, 11, 0.05); border-radius: 0 3px 3px 0;">
  <span style="font-size: 0.65em; font-weight: 700; letter-spacing: 0.13em; text-transform: uppercase; color: #f59e0b;">atenção</span><br>
  <span style="font-size: 0.88em; line-height: 1.65;">Os axiomas de um espaço vetorial não descrevem vetores geométricos — descrevem qualquer conjunto onde adição e multiplicação por escalar façam sentido. Funções contínuas formam um espaço vetorial. Polinômios formam um espaço vetorial. Matrizes formam um espaço vetorial. O vetor de estado de um qubit é um elemento de $\mathbb{C}^2$ — um espaço vetorial sobre $\mathbb{C}$.</span>
</div>

```python
import numpy as np

# ── Os axiomas de espaço vetorial em código ─────────────────────────────────
# Demonstrar que os axiomas valem para vetores em R^n

u = np.array([1.0, 2.0, -1.0])
v = np.array([0.5, -1.0, 3.0])
w = np.array([-2.0, 0.5, 1.5])
a, b = 3.0, -0.5

print('== Axiomas de espaço vetorial ==')
print()
print('1. Comutatividade da adição:    u+v == v+u ?', np.allclose(u+v, v+u))
print('2. Associatividade da adição:   (u+v)+w == u+(v+w) ?', np.allclose((u+v)+w, u+(v+w)))
print('3. Elemento neutro:             u+0 == u ?', np.allclose(u + np.zeros(3), u))
print('4. Elemento inverso:            u+(-u) == 0 ?', np.allclose(u + (-u), np.zeros(3)))
print('5. Distribuição a(u+v):         a*(u+v) == a*u+a*v ?', np.allclose(a*(u+v), a*u+a*v))
print('6. Distribuição (a+b)u:         (a+b)*u == a*u+b*u ?', np.allclose((a+b)*u, a*u+b*u))
print('7. Associatividade escalar:     a*(b*u) == (a*b)*u ?', np.allclose(a*(b*u), (a*b)*u))
print('8. Identidade escalar:          1*u == u ?', np.allclose(1.0*u, u))
print()
print('Os mesmos axiomas valem para polinômios, funções, matrizes, estados quânticos.')
print('O que muda é o que chamamos de "vetor" — a estrutura é a mesma.')
```

## 3. Transformações lineares e matrizes

O segundo grande eixo da álgebra linear não é sobre espaços — é sobre as `transformações` entre espaços.

Uma transformação $T: V \to W$ é **linear** se preserva a estrutura do espaço vetorial:
$$T(\alpha u + \beta v) = \alpha T(u) + \beta T(v)$$

<div style="margin: -0.6em 0 1.4em 0; font-size: 0.87em; color: #6b7280; line-height: 1.65; padding-left: 1.1em; border-left: 2px solid #374151;">
  Preservar a estrutura significa: a transformação de uma combinação linear é a mesma combinação linear das transformações. Isso é uma restrição forte — a maioria das funções não é linear. Mas as que são têm uma propriedade notável: são completamente descritas pelo que fazem aos vetores de uma base.
</div>

Se $V$ tem base $\{e_1, \ldots, e_n\}$ e $W$ tem base $\{f_1, \ldots, f_m\}$, então $T$ é completamente determinada pelos vetores $T(e_1), \ldots, T(e_n)$. Escrever cada $T(e_j)$ nas coordenadas da base de $W$ dá as colunas de uma **matriz** $m \times n$.

Isso é o que Cayley formalizou em 1858: **uma matriz é uma transformação linear disfarçada de tabela de números**.

A composição de transformações corresponde à multiplicação de matrizes — e é por isso que a multiplicação de matrizes tem a fórmula estranha que tem: ela foi projetada para representar composição de funções, não produto número a número.

```python
import numpy as np
import matplotlib.pyplot as plt

# ── Visualização: transformação linear como deformação do plano ──────────────
# Uma matriz 2x2 transforma o plano: veja o que acontece com a grade unitária.

fig, axes = plt.subplots(1, 3, figsize=(14, 5), facecolor='#111')
fig.patch.set_facecolor('#111')

def plot_grade(ax, M, titulo):
    ax.set_facecolor('#111')
    ax.tick_params(colors='#9ca3af')
    for spine in ax.spines.values():
        spine.set_edgecolor('#374151')
    ax.set_aspect('equal')
    ax.axhline(0, color='#374151', lw=0.7)
    ax.axvline(0, color='#374151', lw=0.7)

    # Grade original
    xs = np.linspace(-2, 2, 9)
    for x in xs:
        pontos = np.array([[x, y] for y in np.linspace(-2, 2, 50)]).T
        trans  = M @ pontos
        ax.plot(trans[0], trans[1], color='steelblue', lw=0.7, alpha=0.4)
    for y in xs:
        pontos = np.array([[x, y] for x in np.linspace(-2, 2, 50)]).T
        trans  = M @ pontos
        ax.plot(trans[0], trans[1], color='steelblue', lw=0.7, alpha=0.4)

    # Vetores base transformados
    e1_t = M @ np.array([1, 0])
    e2_t = M @ np.array([0, 1])
    ax.annotate('', xy=e1_t, xytext=(0, 0),
                arrowprops=dict(arrowstyle='->', color='tomato', lw=2.5))
    ax.annotate('', xy=e2_t, xytext=(0, 0),
                arrowprops=dict(arrowstyle='->', color='#a78bfa', lw=2.5))
    ax.text(e1_t[0]+0.1, e1_t[1]+0.1, f'T(e₁)\n{e1_t}', color='tomato', fontsize=8)
    ax.text(e2_t[0]+0.1, e2_t[1]+0.1, f'T(e₂)\n{e2_t}', color='#a78bfa', fontsize=8)

    det = np.linalg.det(M)
    ax.set_title(f'{titulo}\ndet={det:.2f}', color='white', pad=8, fontsize=10)
    ax.set_xlim(-3, 3); ax.set_ylim(-3, 3)

transformacoes = [
    (np.array([[1, 0], [0, 1]]),           'Identidade'),
    (np.array([[2, 1], [0.5, 1.5]]),       'Cisalhamento + escala'),
    (np.array([[0, -1], [1, 0]], dtype=float), 'Rotação 90°'),
]

for ax, (M, titulo) in zip(axes, transformacoes):
    plot_grade(ax, M, titulo)

plt.suptitle('Transformações lineares: cada matriz deforma o plano de forma diferente',
             color='#9ca3af', fontsize=10)
plt.tight_layout()
plt.show()

print('Determinante = fator de escala de área (com sinal).')
print('det=0 → a transformação colapsa o espaço (perde dimensão, não invertível).')
print('det<0 → inverte a orientação.')
```

## 4. Autovalores e autovetores: a estrutura espectral

A maioria das transformações lineares mistura as direções do espaço — um vetor que aponta em $e_1$ pode ser mandado para uma combinação de $e_1$ e $e_2$. Mas algumas direções são especiais: elas não mudam de direção, só de escala.

Um **`autovetor`** de $T$ com **`autovalor`** $\lambda$ é um vetor $v \neq 0$ tal que:
$$T v = \lambda v$$

<div style="margin: -0.6em 0 1.4em 0; font-size: 0.87em; color: #6b7280; line-height: 1.65; padding-left: 1.1em; border-left: 2px solid #374151;">
  Aplicar $T$ a $v$ é o mesmo que multiplicar $v$ por um escalar. A transformação não "gira" $v$ — apenas o estica ou comprime. Os autovetores são as direções invariantes de $T$; os autovalores dizem por quanto.
</div>

O conjunto de autovalores de $T$ é o **`espectro`** de $T$. Para matrizes Hermitianas (operadores de observáveis quânticos), o espectro é sempre real.

A **`decomposição espectral`** diz que uma matriz Hermitiana pode ser escrita como:
$$A = \sum_k \lambda_k |v_k\rangle\langle v_k|$$

Isso significa que $A$ é completamente caracterizada por seus autovetores (as direções invariantes) e seus autovalores (as escalas). Na base de autovetores, $A$ é diagonal — e diagonal é simples.

```python
import numpy as np
import matplotlib.pyplot as plt

# ── Decomposição espectral: visualização geométrica ──────────────────────────
A = np.array([[3.0, 1.0],
              [1.0, 2.0]])

eigvals, eigvecs = np.linalg.eigh(A)   # eigh para matrizes Hermitianas (reais e simétricas)

print('Matriz A:')
print(A)
print()
print(f'Autovalores: λ₁={eigvals[0]:.4f},  λ₂={eigvals[1]:.4f}')
print(f'Autovetor v₁: {eigvecs[:, 0]}')
print(f'Autovetor v₂: {eigvecs[:, 1]}')
print()

# Reconstruir A via decomposição espectral
A_reconstruida = sum(eigvals[k] * np.outer(eigvecs[:, k], eigvecs[:, k]) for k in range(2))
print('A reconstruída via Σ λₖ vₖvₖᵀ:')
print(A_reconstruida)
print('Coincide com A original?', np.allclose(A, A_reconstruida))
print()

# Visualizar: autovetores são as direções invariantes de A
fig, axes = plt.subplots(1, 2, figsize=(12, 5), facecolor='#111')
fig.patch.set_facecolor('#111')
colors_ev = ['tomato', 'steelblue']

for ax in axes:
    ax.set_facecolor('#111')
    ax.tick_params(colors='#9ca3af')
    for spine in ax.spines.values():
        spine.set_edgecolor('#374151')
    ax.set_aspect('equal')
    ax.axhline(0, color='#374151', lw=0.6)
    ax.axvline(0, color='#374151', lw=0.6)
    ax.set_xlim(-3.5, 3.5); ax.set_ylim(-3.5, 3.5)

# Painel esquerdo: vetores antes de A
thetas = np.linspace(0, 2*np.pi, 24, endpoint=False)
for theta in thetas:
    v = np.array([np.cos(theta), np.sin(theta)])
    axes[0].annotate('', xy=v, xytext=(0, 0),
                     arrowprops=dict(arrowstyle='->', color='#374151', lw=1))
for i, (v, color) in enumerate(zip(eigvecs.T, colors_ev)):
    axes[0].annotate('', xy=v, xytext=(0, 0),
                     arrowprops=dict(arrowstyle='->', color=color, lw=2.5))
    axes[0].text(v[0]+0.1, v[1]+0.1, f'v{i+1}', color=color, fontsize=10, fontweight='bold')
axes[0].set_title('Vetores unitários (círculo)\nAutovetores em destaque', color='white', pad=10)

# Painel direito: vetores após A — autovetores só escalam
for theta in thetas:
    v = np.array([np.cos(theta), np.sin(theta)])
    Av = A @ v
    axes[1].annotate('', xy=Av, xytext=(0, 0),
                     arrowprops=dict(arrowstyle='->', color='#374151', lw=1))
for i, (v, lam, color) in enumerate(zip(eigvecs.T, eigvals, colors_ev)):
    Av = A @ v   # = λv
    axes[1].annotate('', xy=Av, xytext=(0, 0),
                     arrowprops=dict(arrowstyle='->', color=color, lw=2.5))
    axes[1].text(Av[0]+0.1, Av[1]+0.1, f'Av{i+1} = {lam:.2f}v{i+1}', color=color, fontsize=9)
axes[1].set_title('Após aplicar A\nAutovetores apenas escalados por λ', color='white', pad=10)

plt.suptitle('Autovetores: as únicas direções que A não rotaciona', color='#9ca3af', fontsize=10)
plt.tight_layout()
plt.show()
```

## 5. Álgebra multilinear: quando um índice não é suficiente

Vetores têm um índice: $v_i$. Matrizes têm dois: $M_{ij}$. Tensores generalizam: um tensor de `rank` $r$ tem $r$ índices.

Mas o que é um tensor, exatamente? Não é apenas uma tabela de números com muitos índices — é uma tabela de números que se transforma de forma específica quando você muda de base.

A definição formal: um **`tensor`** de tipo $(p, q)$ é um objeto que, em cada base, é representado por $n^{p+q}$ números, e que se transforma sob mudança de base de acordo com a regra:

$$T'^{i_1 \ldots i_p}_{j_1 \ldots j_q} = \sum \frac{\partial x'^{i_1}}{\partial x^{k_1}} \cdots \frac{\partial x^{l_1}}{\partial x'^{j_1}} \cdots T^{k_1 \ldots k_p}_{l_1 \ldots l_q}$$

<div style="margin: -0.6em 0 1.4em 0; font-size: 0.87em; color: #6b7280; line-height: 1.65; padding-left: 1.1em; border-left: 2px solid #374151;">
  Um escalar é um tensor de rank 0 — nenhum índice. Um vetor é rank 1 — um índice. Uma matriz é rank 2 — dois índices. O que muda não é o número de índices em si, mas a lei de transformação: um tensor verdadeiro transforma-se de forma covariante com a mudança de base, garantindo que as equações físicas sejam independentes do sistema de coordenadas.
</div>

A motivação histórica era geométrica: Ricci (1887) e Levi-Civita precisavam de objetos que descrevessem curvatura do espaço de forma independente das coordenadas. Einstein (1915) usou exatamente esse formalismo para a relatividade geral — o tensor métrico $g_{\mu\nu}$ descreve como distâncias se medem em cada ponto do espaço-tempo curvo.

## 6. O produto tensorial: construindo espaços maiores

Além dos tensores como objetos com múltiplos índices, há uma operação que constrói novos espaços a partir de espaços dados: o **`produto tensorial`** $V \otimes W$.

Dados dois espaços vetoriais $V$ (dimensão $m$) e $W$ (dimensão $n$), o produto tensorial $V \otimes W$ é um espaço de dimensão $mn$. Seus elementos são combinações lineares de produtos tensoriais de vetores: $v \otimes w$, onde $v \in V$ e $w \in W$.

A regra de construção:
$$\dim(V \otimes W) = \dim(V) \cdot \dim(W)$$

Se $\{e_1, \ldots, e_m\}$ é base de $V$ e $\{f_1, \ldots, f_n\}$ é base de $W$, então $\{e_i \otimes f_j\}$ é base de $V \otimes W$ — com $mn$ vetores.

<div style="border-left: 2px solid #f59e0b; padding: 0.5em 1em 0.5em 1.1em; margin: 0.8em 0 1.2em 0; background: rgba(245, 158, 11, 0.05); border-radius: 0 3px 3px 0;">
  <span style="font-size: 0.65em; font-weight: 700; letter-spacing: 0.13em; text-transform: uppercase; color: #f59e0b;">atenção</span><br>
  <span style="font-size: 0.88em; line-height: 1.65;">Nem todo elemento de $V \otimes W$ pode ser escrito como $v \otimes w$ para vetores individuais $v$ e $w$. Elementos que <em>podem</em> ser escritos assim são chamados de <b>produtos tensoriais</b> ou <b>estados separáveis</b>. Os que <em>não podem</em> são <b>estados emaranhados</b>. O emaranhamento quântico é, matematicamente, a não-separabilidade no produto tensorial.</span>
</div>

```python
import numpy as np

# ── Produto tensorial de vetores (produto de Kronecker) ──────────────────────

ket_0 = np.array([1, 0], dtype=complex)   # |0⟩
ket_1 = np.array([0, 1], dtype=complex)   # |1⟩
ket_p = np.array([1, 1], dtype=complex) / np.sqrt(2)   # |+⟩

# Estado separável: |0⟩ ⊗ |1⟩ = |01⟩
estado_01 = np.kron(ket_0, ket_1)
print('|0⟩ ⊗ |1⟩ =', estado_01)   # [0, 1, 0, 0]

# Estado separável: |+⟩ ⊗ |+⟩
estado_pp = np.kron(ket_p, ket_p)
print('|+⟩ ⊗ |+⟩ =', np.round(estado_pp, 4))
print()

# Estado de Bell: 1/√2 (|00⟩ + |11⟩) — NÃO é produto tensorial de dois qubits
bell = (np.kron(ket_0, ket_0) + np.kron(ket_1, ket_1)) / np.sqrt(2)
print('Estado de Bell: (|00⟩+|11⟩)/√2 =', np.round(bell, 4))
print()

# ── Verificar se o estado de Bell é separável ────────────────────────────────
# Um estado |ψ⟩ ∈ C^4 é separável ↔ pode ser escrito como α⊗β
# Equivalente: a matriz 2x2 obtida por reshape tem rank 1

def rank_separabilidade(estado):
    """Retorna o rank da matriz de Schmidt — 1 se separável, >1 se emaranhado."""
    M = estado.reshape(2, 2)
    singular_values = np.linalg.svd(M, compute_uv=False)
    rank = np.sum(singular_values > 1e-10)
    return rank, singular_values

for nome, estado in [('|01⟩', estado_01), ('|+⟩⊗|+⟩', estado_pp), ('Bell', bell)]:
    rank, sv = rank_separabilidade(estado)
    separavel = 'separável' if rank == 1 else 'EMARANHADO'
    print(f'{nome:12s} → rank de Schmidt = {rank}  ({separavel})  sv={np.round(sv, 3)}')
```

## 7. O `rank de Schmidt` e a decomposição de Schmidt

Para um estado bipartido $|\psi\rangle \in \mathcal{H}_A \otimes \mathcal{H}_B$, a **`decomposição de Schmidt`** é a expansão:

$$|\psi\rangle = \sum_{k=1}^{r} \lambda_k |a_k\rangle \otimes |b_k\rangle$$

onde $\lambda_k > 0$ são os **coeficientes de Schmidt** e $r$ é o **rank de Schmidt**.

- $r = 1$: estado `separável` — existe $|a\rangle$ e $|b\rangle$ tais que $|\psi\rangle = |a\rangle \otimes |b\rangle$
- $r > 1$: estado **emaranhado** — não pode ser fatorado

A decomposição de Schmidt é, na prática, a `decomposição em valores singulares (SVD)` da matriz obtida pelo reshape do vetor de estado. Isso conecta emaranhamento quântico com a álgebra linear mais clássica.

<a href="../digressoes/D-02.ipynb" style="font-size: 0.82em; color: #6b7280; text-decoration: none; border-bottom: 1px dashed #6b7280; white-space: nowrap;">↳ D-02 — Produtos Tensoriais: como combinar estados de múltiplos qubits</a>

```python
import numpy as np

# ── Decomposição de Schmidt via SVD ──────────────────────────────────────────

def decomposicao_schmidt(estado):
    """Decompõe |ψ⟩ ∈ C^(nA*nB) em Σ λk |ak⟩⊗|bk⟩ via SVD."""
    nA = nB = int(np.sqrt(len(estado)))   # assume nA = nB
    M = estado.reshape(nA, nB)
    U, s, Vh = np.linalg.svd(M)
    # Coeficientes não-nulos
    r = np.sum(s > 1e-10)
    lambdas = s[:r]
    kets_A  = [U[:, k] for k in range(r)]
    kets_B  = [Vh[k, :] for k in range(r)]
    return lambdas, kets_A, kets_B

ket_0 = np.array([1, 0], dtype=complex)
ket_1 = np.array([0, 1], dtype=complex)
bell  = (np.kron(ket_0, ket_0) + np.kron(ket_1, ket_1)) / np.sqrt(2)

lambdas, kets_A, kets_B = decomposicao_schmidt(bell)

print('Estado de Bell: (|00⟩ + |11⟩)/√2')
print()
print(f'Rank de Schmidt: {len(lambdas)}')
print()
for k, (lam, a, b) in enumerate(zip(lambdas, kets_A, kets_B)):
    print(f'  λ{k+1} = {lam:.4f}')
    print(f'  |a{k+1}⟩ = {np.round(a, 4)}')
    print(f'  |b{k+1}⟩ = {np.round(b, 4)}')
    print()

print('Verificação: Σ λk |ak⟩⊗|bk⟩ =')
reconstruido = sum(lam * np.kron(a, b) for lam, a, b in zip(lambdas, kets_A, kets_B))
print(np.round(reconstruido, 4))
print('Coincide com bell?', np.allclose(reconstruido, bell))
```

## Por que isso importa para algoritmos quânticos

A trajetória al-Khwārizmī → Galois → Cayley → Ricci não é história por curiosidade — ela é a fundação estrutural de como qubits e algoritmos funcionam:

**Espaços vetoriais:** o estado de $n$ qubits vive em $\mathbb{C}^{2^n} = \mathbb{C}^2 \otimes \cdots \otimes \mathbb{C}^2$ — produto tensorial de $n$ espaços de um qubit. A dimensão cresce exponencialmente: 10 qubits = 1024 dimensões, 50 qubits = $10^{15}$ dimensões. Isso é a fonte da capacidade computacional e do custo de simulação clássica.

**Transformações lineares:** cada porta quântica é uma matriz unitária — uma transformação linear que preserva a norma. Circuitos quânticos são composições de transformações lineares.

**Espectro:** QPE extrai autovalores. Shor explora a estrutura espectral de $f(a) = x^a \bmod N$. VQE minimiza $\langle \psi | \hat{H} | \psi \rangle$. Toda a computação quântica útil é, no fundo, processamento espectral.

**Produto tensorial e emaranhamento:** o emaranhamento — o recurso computacional específico da computação quântica sem análogo clássico — é matematicamente a não-separabilidade no produto tensorial. O rank de Schmidt quantifica o grau de emaranhamento via SVD.

**Tensores de rank alto:** algoritmos como QSP/QSVT manipulam polinômios de operadores — objetos de estrutura tensorial complexa. Entender tensores é entender o espaço onde a computação quântica avançada opera.
