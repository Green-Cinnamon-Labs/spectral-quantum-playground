## O problema que motivou tudo

*Esqueleto — expandir aqui*

Ideia central a desenvolver:

- Fourier (1822) queria descrever como o calor se propaga num sólido. A equação do calor tinha soluções que pareciam senoidais — mas funções arbitrárias não são senoidais. A pergunta era: qualquer função pode ser escrita como soma de oscilações?
- Isso era filosoficamente controverso na época (Lagrange objetou). Mas a ideia sobreviveu porque funcionava.
- O ponto não é só elegância matemática: decompor em frequências separa escalas. Processar cada frequência independentemente e depois recompor é muito mais barato do que processar o sinal inteiro de uma vez.
- Trajetória desta seção: da motivação física à pergunta matemática central — "quanto de cada frequência existe neste sinal?".

## A Série de Fourier

*Esqueleto — expandir aqui*

Ideia central a desenvolver:

- Funções periódicas contínuas com período $T$ podem ser escritas como:
$$f(t) = \sum_{k=-\infty}^{\infty} c_k \, e^{i 2\pi k t / T}$$
- Os coeficientes $c_k$ são o "quanto" de cada frequência $k/T$ existe em $f$:
$$c_k = \frac{1}{T} \int_0^T f(t) \, e^{-i 2\pi k t / T} \, dt$$
- Por que exponenciais complexas e não senos e cossenos? Elas formam uma base ortonormal no espaço de funções de quadrado integrável — o produto interno natural aqui é exatamente $\langle f, g \rangle = \int_0^T f(t)\overline{g(t)}\,dt$. Conectar com M-00: os $e^{i2\pi kt/T}$ são os vetores de base; os $c_k$ são as coordenadas.
- Mostrar: ortogonalidade das exponenciais — $\langle e^{i2\pi k t/T}, e^{i2\pi j t/T}\rangle = \delta_{kj}$.
- Intuição geométrica: projeção de $f$ nas direções de oscilação pura.

## De funções para sequências: a DFT

*Esqueleto — expandir aqui*

Ideia central a desenvolver:

- No mundo digital não temos funções contínuas. Temos $N$ amostras: $x_0, x_1, \ldots, x_{N-1}$.
- A **Transformada de Fourier Discreta (DFT)** mapeia essas $N$ amostras em $N$ coeficientes complexos:
$$X_k = \sum_{j=0}^{N-1} x_j \, \omega_N^{-jk}, \quad \omega_N = e^{i2\pi/N}$$
- Inversa:
$$x_j = \frac{1}{N} \sum_{k=0}^{N-1} X_k \, \omega_N^{jk}$$
- O que $X_k$ mede: a correlação da sequência $x$ com a oscilação de frequência $k/N$.

## A DFT é uma transformação linear

*Esqueleto — expandir aqui*

Ideia central a desenvolver:

- A DFT pode ser escrita como multiplicação por uma matriz $F_N$:
$$\vec{X} = F_N \vec{x}, \quad (F_N)_{kj} = \frac{1}{\sqrt{N}} \omega_N^{jk}$$
- $F_N$ é **unitária**: $F_N^\dagger F_N = I$. Conectar com M-00: transformação linear que preserva norma.

## A DFT encontra periodicidade

*Esqueleto — expandir aqui*

Ideia central a desenvolver — esta é a seção mais importante para o projeto:

- Se $x$ tem período $r$ (com $r | N$), então $X_k \neq 0$ somente quando $k$ é múltiplo de $N/r$. A DFT concentra toda a energia nesses $r$ frequências.
- Por quê? As $N/r$ cópias do período somam construtivamente exatamente nas frequências $0, N/r, 2N/r, \ldots$ — interferência construtiva. Em todas as outras frequências, cancelamento completo.
- Mostrar com exemplo: $N=8$, sequência periódica com $r=4$. Plotar espectro.
- Intuição: o espectro de uma função periódica é discreto — os "picos" estão nos harmônicos do período fundamental.
- Esta propriedade é exatamente o que o algoritmo de Shor explora: se construirmos uma sequência $f(a) = x^a \bmod N$ e aplicarmos a DFT, os picos revelam o período $r$ de $f$. E a fatoração de $N$ reduz ao problema de encontrar $r$.

## A Transformada de Fourier Quântica (QFT)

*Esqueleto — expandir aqui*

Ideia central a desenvolver:

- A QFT é a **mesma matriz $F_N$**, com $N = 2^n$, implementada como circuito quântico em $n$ qubits.
- Ação: $\text{QFT}|j\rangle = \frac{1}{\sqrt{N}} \sum_{k=0}^{N-1} \omega_N^{jk} |k\rangle$
- Em superposição: $\text{QFT}\left(\sum_j x_j |j\rangle\right) = \sum_k X_k |k\rangle$ — aplica a DFT sobre as amplitudes.
- O circuito usa $O(n^2)$ portas (Hadamard + rotações de fase controladas), contra $O(N \log N) = O(2^n \cdot n)$ clássico. Speedup exponencial em construir a transformada.
- **Mas há um custo:** não se pode ler todos os $X_k$ — a medição colapsa para um único $k$ com probabilidade $|X_k|^2$. Usamos isso estatisticamente, não diretamente.
- Mostrar o circuito para $n=3$ ($N=8$): estrutura de Hadamards e portas $R_k = \text{diag}(1, e^{i2\pi/2^k})$.

## O que você pode — e não pode — ler depois da QFT

*Esqueleto — expandir aqui*

Ideia central a desenvolver — esta seção é sobre os limites, essencial para honestidade:

- Após QFT, o estado é $\sum_k X_k |k\rangle$. Medir dá um único $k$ com probabilidade $|X_k|^2$.
- Se o espectro é plano (energia distribuída uniformemente), uma medição não diz nada útil.
- Se o espectro tem picos (sinal periódico), medições repetidas amostram os harmônicos. Com frações contínuas, $r$ pode ser recuperado de uma única medição com alta probabilidade.
- Portanto: a QFT é útil quando o espectro é **esparso** — poucos picos concentram a probabilidade.
- Conectar com QPE: a fase $\phi$ de um autovalor $e^{i2\pi\phi}$ cria exatamente um sinal periódico no registrador de controle. A QFT inversa então concentra a probabilidade em $|\tilde{\phi}\rangle$.

## Conexão com o projeto

*Esqueleto — expandir aqui*

Três fios a conectar:

**QFT e QPE (C-04):** No algoritmo de estimação de fase, após os kickbacks de fase, o registrador de controle está em $\sum_k e^{i2\pi k \phi} |k\rangle / \sqrt{N}$ — exatamente a DFT de uma função periódica com período $1/\phi$. A QFT inversa concentra a amplitude em $|\tilde{\phi}\rangle$. A QFT é o mecanismo de decodificação espectral.

**QFT e Shor (C-06):** A sequência $f(a) = x^a \bmod N$ é periódica com período $r$ (a ordem de $x$ módulo $N$). Preparar a superposição $\sum_a |a\rangle|f(a)\rangle$ e aplicar a QFT no primeiro registrador revela $r$ nos picos de probabilidade. Shor é: fatoração → encontrar $r$ → QFT.

**Hadamard como QFT de 1 qubit:** $H = F_2$. O circuito mais simples de computação quântica já é uma Transformada de Fourier. Isso não é coincidência — a razão pela qual $H$ cria superposição uniforme é que ela distribui energia igualmente na base de Fourier, que é a base dos estados de frequência definida.

Pontos a cobrir no resumo:

- A série de Fourier decompõe funções periódicas em somas de oscilações puras; os coeficientes são projeções em vetores de base ortonormais — o produto interno do espaço de funções.
- A DFT faz o mesmo para sequências finitas: mapeia $N$ amostras em $N$ coeficientes de frequência via multiplicação pela matriz unitária $F_N$.
- Uma sequência periódica com período $r$ produz espectro esparso: energia concentrada nos harmônicos $k = 0, N/r, 2N/r, \ldots$
- A QFT implementa $F_N$ como circuito quântico em $O(n^2)$ portas — speedup exponencial. Mas só permite amostragem do espectro, não leitura completa.
- A QFT é útil exatamente quando o espectro tem estrutura (picos) — sinais periódicos, fases de autovalores.
- A porta Hadamard é $F_2$: toda a computação quântica começa com uma Transformada de Fourier.
