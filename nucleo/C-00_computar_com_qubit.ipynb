## Pergunta

Por que é possível computar com um qubit? *(→ [o que é um qubit](../digressoes/D-00_o_que_e_um_qubit.ipynb))*

> “Computar” vem do latim computare: com- (“junto”, “em conjunto”) + putare (“calcular”, “avaliar”, “contar”, e originalmente até “podar”, no sentido de limpar para apurar). A ideia original é a de contar ou calcular reunindo elementos. Depois, o termo ganhou o sentido mais técnico de processar informação segundo regras. Então, no fundo, “computar” carrega a ideia de transformar algo em resultado por meio de um procedimento bem definido.

Eu consigo computar com um qubit porque, depois de prepará-lo em um certo estado e submetê-lo a uma evolução controlada, a própria mecânica quântica transforma esse estado até que, na medição, um resultado seja obtido.

Este notebook constrói essa intuição antes de qualquer formalismo.

## 1. Apesar de envolver probabilidade, não é sobre probabilidade

Na prática, computar com um qubit é manipular amplitude e fase. É isso que está sendo controlado fisicamente. No fim, a medição converte essa estrutura em probabilidades observáveis. A probabilidade é só o que aparece no fim, na leitura; o que realmente exploramos durante o processo é a interferência. Em outras palavras: o ganho quântico não vem de “ter probabilidades”, porque sistemas clássicos também têm isso; ele vem de poder manipular amplitudes e fase de modo que algumas possibilidades se reforcem e outras se cancelem antes da medição.

Dado um problema prático, o primeiro passo é encontrar uma forma de representá-lo dentro de um sistema quântico. Feito isso, podemos usar a interferência para transformar essa representação ao longo do tempo, reforçando os estados associados às boas respostas e enfraquecendo os demais. O ponto importante é este: o sistema não “descobre” a resposta por mágica; ele foi cuidadosamente preparado para evoluir de um jeito que favoreça a leitura da resposta desejada.

## 2. O experimento mais simples possível

Considere o circuito de um qubit:

$$|0\rangle \xrightarrow{H} |{+}\rangle \xrightarrow{Z} |{-}\rangle \xrightarrow{H} |1\rangle$$

Resultado: $H \cdot Z \cdot H = X$ (bit flip determinístico).

Por quê? $H$ cria superposição, $Z$ inverte a fase de $|1\rangle$, $H$ fecha a interferência. O resultado é determinístico — não é sorte, é consequência estrutural das fases.

Um computador probabilístico clássico não pode fazer isso com um único "bit probabilístico": sem amplitudes complexas, não há mecanismo para a interferência destrutiva que cancela $|0\rangle$ no final.

```python
import numpy as np

# Portas elementares
H = np.array([[1, 1], [1, -1]], dtype=complex) / np.sqrt(2)
Z = np.array([[1, 0], [0, -1]], dtype=complex)
X = np.array([[0, 1], [1, 0]], dtype=complex)

ket_0 = np.array([1.0, 0.0], dtype=complex)

# Circuito H -> Z -> H
after_H   = H @ ket_0
after_Z   = Z @ after_H
after_H2  = H @ after_Z

print('|0>      :', ket_0)
print('H|0>     :', after_H)
print('ZH|0>    :', after_Z)
print('HZH|0>   :', after_H2)
print()
print('X|0>     :', X @ ket_0)
print()
print('HZH == X?', np.allclose(H @ Z @ H, X))
```

## 3. O que a interferência fez aqui

Acompanhando as amplitudes explicitamente:

$$|0\rangle = \begin{pmatrix}1\\0\end{pmatrix}$$

$$H|0\rangle = \frac{1}{\sqrt{2}}\begin{pmatrix}1\\1\end{pmatrix}$$

$$ZH|0\rangle = \frac{1}{\sqrt{2}}\begin{pmatrix}1\\-1\end{pmatrix}$$

$$HZH|0\rangle = \frac{1}{2}\begin{pmatrix}1-1\\1+1\end{pmatrix} = \begin{pmatrix}0\\1\end{pmatrix} = |1\rangle$$

O $Z$ inverteu a fase relativa. O segundo $H$ combinou as amplitudes de forma que $|0\rangle$ se cancelou (*interferência destrutiva*) e $|1\rangle$ se reforçou (*interferência construtiva*). Resultado determinístico.

```python
import matplotlib.pyplot as plt

steps = {
    '|0⟩': ket_0,
    'H|0⟩': after_H,
    'ZH|0⟩': after_Z,
    'HZH|0⟩': after_H2,
}

fig, axes = plt.subplots(1, 4, figsize=(12, 3))
for ax, (label, state) in zip(axes, steps.items()):
    probs = np.abs(state)**2
    bars = ax.bar(['|0⟩', '|1⟩'], probs, color=['steelblue', 'tomato'])
    ax.set_title(label)
    ax.set_ylim(0, 1.1)
    ax.set_ylabel('$|\\mathrm{amp}|^2$')
    for bar, amp in zip(bars, state):
        ax.text(bar.get_x() + bar.get_width()/2, bar.get_height() + 0.03,
                f'{amp.real:+.2f}', ha='center', va='bottom', fontsize=9)

plt.suptitle('Amplitudes (topo de cada barra) e probabilidades (altura) — circuito H→Z→H',
             fontsize=10)
plt.tight_layout()
plt.show()
```

## 4. Por que isso não é possível com probabilidades

Esboço: num modelo probabilístico clássico o estado é $(p_0, p_1)$ com $p_i \geq 0$. Qualquer operação estocástica preserva $p_i \geq 0$. Não há cancelamento — interferência destrutiva requer somar amplitudes com sinais opostos, o que não existe no espaço de probabilidades.
