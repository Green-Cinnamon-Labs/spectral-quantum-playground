# N-03 — Como eu posso medir o resultado?

## Afirmação introdutória


A medição extrai um resultado observável, com probabilidades dadas pela `regra de Born`, e substitui a superposição pelo estado medido.


## 1. O que a Medição faz ao 'Vetor Estado' — a Regra de Born

Computar com um qubit é preparar um estado inicial, aplicar transformações sobre ele e só então medir. Durante essa evolução, o qubit é descrito por um vetor de estado, e as operações unitárias vão mudando esse vetor: redistribuem amplitudes, alteram fases e exploram interferência. O objetivo desse processo não é “ver a superposição”, mas usar essa superposição como matéria-prima para reorganizar as probabilidades de forma controlada.

A medição entra no fim. Quando você mede na base computacional, o estado deixa de ser a combinação anterior e produz um resultado clássico: $0$ ou $1$. Esse resultado não surge do nada; ele reflete tudo o que foi feito antes pelas transformações. Em outras palavras: a computação quântica trabalha no estado vetorial durante o processo, e a medição converte esse processo em uma resposta observável. O desafio do algoritmo é justamente fazer com que, no momento da medição, a resposta relevante para o problema tenha alta chance de aparecer.
