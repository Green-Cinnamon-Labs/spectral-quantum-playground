## O que é

*Esqueleto — expandir aqui*

Ideia central a desenvolver:

- Dan Simon (1994) propôs um problema que, à primeira vista, parece abstrato: dada uma função $f:\{0,1\}^n \to \{0,1\}^n$ que só pode ser consultada como oráculo, encontre um vetor secreto $s \in \{0,1\}^n$ tal que $f(x) = f(y) \iff x \oplus y \in \{0^n, s\}$.
- Em palavras: $f$ é periódica sobre o grupo $\{0,1\}^n$ com período $s$. Se $s = 0^n$, $f$ é injetiva. Caso contrário, cada valor é atingido exatamente duas vezes — por $x$ e por $x \oplus s$.
- Classicamente: sem sorte, você precisa $\Omega(2^{n/2})$ consultas ao oráculo para encontrar $s$ com alta probabilidade (aniversário).
- Quanticamente: $O(n)$ consultas bastam.

## A solução quântica

*Esqueleto — expandir aqui*

Ideia central a desenvolver:

1. **Preparação:** $|0^n\rangle|0^n\rangle$
2. **Superposição:** aplica $H^{\otimes n}$ no primeiro registrador → $\frac{1}{\sqrt{2^n}}\sum_x |x\rangle|0^n\rangle$
3. **Oráculo:** $U_f$ → $\frac{1}{\sqrt{2^n}}\sum_x |x\rangle|f(x)\rangle$
4. **Medir o segundo registrador:** colapsa para algum valor $f(x_0)$. O primeiro registrador colapsa para $\frac{1}{\sqrt{2}}(|x_0\rangle + |x_0 \oplus s\rangle)$.
5. **Hadamard no primeiro registrador:** produz um vetor $y$ uniformemente distribuído sobre todos os $y$ com $y \cdot s = 0 \pmod{2}$.
6. **Medir:** obter um $y$ que satisfaz $y \cdot s = 0$.
7. **Repetir $O(n)$ vezes:** acumular $n-1$ equações linearmente independentes → resolver o sistema linear sobre $\mathbb{F}_2$ para obter $s$.

O passo 5 é uma **Transformada de Fourier sobre $\{0,1\}^n$** — exatamente o que Simon reconheceu e que Shor generalizou.

## Por que importa aqui

*Esqueleto — expandir aqui*

Ideia central a desenvolver:

- Simon é o **caso mais simples de detecção de estrutura escondida por Fourier quântico**. A estrutura aqui é um subgrupo de $\{0,1\}^n$ com gerador $s$. A Transformada de Hadamard (que é $F_2^{\otimes n}$, a QFT sobre o grupo $\mathbb{Z}_2^n$) produz saídas ortogonais ao período.
- Shor é Simon com o grupo trocado: em vez de $\mathbb{Z}_2^n$ (hipercubo binário), trabalha sobre $\mathbb{Z}_N$ (inteiros módulo $N$). A QFT sobre $\mathbb{Z}_N$ substitui o Hadamard. A ideia é a mesma: transformada de Fourier revela o período.
- A palestra de Shor (1994) deixa explícito: *"Eu olhei para o problema de Simon e sabia que transformadas de Fourier eram muito boas para encontrar periodicidade."*
- Entender Simon é entender o núcleo estrutural do algoritmo de Shor sem a complexidade da aritmética modular.
