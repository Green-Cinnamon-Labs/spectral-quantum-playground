## O que é

Bom, primeiro vou fizer o que é formalmente e depois vou traduzir.

> Um `qubit` é qualquer sistema físico de dois níveis que obedece à mecânica quântica. Matematicamente, é um vetor normalizado em $\mathbb{C}^2$ — um espaço de Hilbert, *(→ [vetores, Espaços e Espaço de Hilbert](../matematica/M-00_vetores_espaco_de_hilbert.ipynb))* — que respeita:

$$|\psi\rangle = \alpha|0\rangle + \beta|1\rangle, \quad \alpha, \beta \in \mathbb{C}, \quad |\alpha|^2 + |\beta|^2 = 1$$

<div style="max-width: 720px; margin: 0 auto; text-align: center; font-size: 0.9em; color: #aaa; line-height: 1.5;">
  A ideia aqui é que o estado quântico pode ser escrito como uma combinação de dois estados básicos, |0⟩ e |1⟩. Os coeficientes α e β dizem "quanto" de cada possibilidade existe nesse estado, mas não como números comuns: eles são amplitudes complexas. Ao medir, essas amplitudes se convertem em probabilidades, e por isso |α|² + |β|² precisa ser igual a 1.
</div>

Dizer “um sistema físico” é só uma forma mais formal de dizer “uma coisa real do mundo”, algo palpável, que existe de verdade. E no caso, este sistema seria especial porque manifesta-se de acordo com as regras, com as leis, da mecânica quântica. 

Exemplos físicos: 
- spin de elétron ($\uparrow$/$\downarrow$)
- polarização de fóton (horizontal/vertical)
- dois níveis de energia de átomo. 

Todos são objetos físicos e que se manifestam através de uma mesma matemática, a mecânica quântica. Acima, você enxerga uma expressão que é bastante avançada. O símbolo $\psi\rangle$ expressa um objeto matemático especial chamado `vetor de estado`, ele é um vetor em um espaço complexo específico conhecido como `espaço de Hilbert`. O que tem de especial nesse vetor? Ele foi escolhido para armazenar toda a informação do sistema em um dado instante.

## Qual tipo de informação esse Vetor está guardando?

Eu acho que essa coisa de "toda a informação do sistema em um dado instante" é confusa. Veja, estamos falando de **mecânica quântica**. 

- **Mecânica** vem do grego mēkhanikḗ, ligado a mēkhanḗ, que significava “máquina”, “engenho”, “artifício”; depois, em latim, virou mechanica, sempre com a ideia de estudar movimento, forças e funcionamento dos corpos. 
- **Quântica** vem do latim quantus (“quanto”, “quanto de quantidade”), e entrou na física por causa da noção de quantum: uma quantidade discreta, um “pacote” mínimo. 
   
Então, a noção aqui é que estamos querendo descrever, estudar, o movimento de sistemas físicos em escala "quântica". Nessa escala, as leis clássicas de movimento perdem o valor e outras começam a valer. Se você pergunta "qual é a trajetória dessa partícula?", você receberá uma resposta probabilística. Com efeito, um dos princípios da MQ diz que não é possível determinar simultaneamente, com precisão arbitrária, a posição e o momento linear de uma partícula. É o famoso Princípio da Incerteza de Heisenberg, expresso por:

$$
\Delta x \, \Delta p \geq \frac{\hbar}{2}
$$

<div style="max-width: 720px; margin: 0 auto; text-align: center; font-size: 0.9em; color: #aaa; line-height: 1.5;">
  A ideia é que esse produto nunca pode cair abaixo de um limite mínimo, dado por ħ/2. Em outras palavras: se tentamos reduzir demais a incerteza de uma grandeza, a da outra precisa crescer de modo a manter esse limite.
</div>

> Quando digo que não é possível, quero dizer que a própria natureza não permite isso. Não é como se eu estivesse dizendo "nossos equipamentos não conseguem"; estou dizendo que a realidade não deixa. Não importa quão poderoso seja o instrumento: nem mesmo um observador perfeito — um anjo, ou até Deus — seria capaz de determinar isso com precisão arbitrária. A realidade, nesse nível, é feita para ser assim.

Bom, dito isso, se você sabe algo sobre esse sistema quântico, esse algo é uma mera probabilidade. Um mapa de possibilidades. Ele nos diz de quais maneiras esse sistema pode se manifestar **quando for medido**, com que probabilidades isso pode acontecer, e como essas possibilidades se relacionam entre si.

## Por que importa aqui

Richard Feynman teve uma intuição importante e que responde porque isso importa. Na década de 80, muitos pesquisadores estavam interessados em um método que permitisse simular como moléculas e processos químicos evoluem no tempo. A questão é que essa evolução é justamente determinada pela mecânica quântica. Descrever essa evolução num computador clássico fica rapidamente inviável do ponto de vista de memória e poder computacional, levaria muito tempo.

Feynman pensou algo assim: "e se, em vez de simular um sistema quântico com uma máquina clássica, eu usar outro sistema quântico controlável e o deixar evoluir segundo a mesma dinâmica para depois medi-lo?". Então, começaram uma busca por um sistema físico simples, controlável e legível que também obedeça à mecânica quântica. Só existem três exigências:

1. Ser preparável em um estado inicial conhecido;
2. Ser manipulável de forma previsível ao longo do tempo;
3. Ser medível no fim, sem ambiguidade;

Ou seja, não serve qualquer sistema quântico. Ele precisa ter estados distingíveis o bastante para a leitura, mas ainda preservar a dinâmica quântica durante sua evolução. Acredito que, pela influência da computação clássica, pensaram em um sistema de dois níveis possíveis porque ele é a forma mais simples de representar e manipular informação fisicamente. Se eu consigo distinguir dois estados, prepará-los de modo confiável, deixá-los evoluir e depois medi-los, então já tenho um objeto físico minimamente útil. 

A sacada envolvendo a mecânica quântica foi perceber que, nesse caso, esses dois níveis não precisavam se comportar como um bit comum: eles podiam evoluir como um vetor de estado, carregando amplitudes, fase e interferência. Então a escolha de um sistema de dois níveis não veio porque a natureza exige isso, mas porque esse era o ponto mais simples e controlável a partir do qual se podia começar a construir uma teoria de manipulação da informação quântica.

## Prática para Visualizar

Vamos passar pelos três momentos fundamentais da vida de um qubit: **preparação**, **operação** e **leitura**. A visualização usa a **Esfera de Bloch** — uma representação geométrica onde qualquer estado de um qubit é um ponto na superfície de uma esfera unitária.

- **Polo norte** → |0⟩  
- **Polo sul** → |1⟩  
- **Qualquer outro ponto** → superposição de |0⟩ e |1⟩, com amplitudes e fase

A operação que vamos aplicar é a **porta Hadamard** — ela move o vetor do polo norte para o equador, criando uma superposição perfeita. Depois, uma **medição** força o vetor a colapsar para um dos polos.

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

def state_to_bloch(alpha, beta):
    """Converte amplitudes (α, β) para coordenadas (x, y, z) na esfera de Bloch."""
    norm = np.sqrt(abs(alpha)**2 + abs(beta)**2)
    a, b = alpha / norm, beta / norm
    theta = 2 * np.arccos(np.clip(abs(a), 0, 1))
    phi = np.angle(b) - np.angle(a)
    return np.array([
        np.sin(theta) * np.cos(phi),
        np.sin(theta) * np.sin(phi),
        np.cos(theta)
    ])

def draw_bloch(ax, bloch_vec, title="", color='crimson'):
    u = np.linspace(0, 2 * np.pi, 40)
    v = np.linspace(0, np.pi, 40)
    sx = np.outer(np.cos(u), np.sin(v))
    sy = np.outer(np.sin(u), np.sin(v))
    sz = np.outer(np.ones_like(u), np.cos(v))
    ax.plot_wireframe(sx, sy, sz, alpha=0.08, color='steelblue', linewidth=0.5)

    t = np.linspace(0, 2 * np.pi, 100)
    ax.plot(np.cos(t), np.sin(t), np.zeros_like(t), 'gray', alpha=0.3, lw=0.8)
    ax.plot(np.cos(t), np.zeros_like(t), np.sin(t), 'gray', alpha=0.3, lw=0.8)
    ax.plot(np.zeros_like(t), np.cos(t), np.sin(t), 'gray', alpha=0.3, lw=0.8)

    for end in [[1.2,0,0],[0,1.2,0],[0,0,1.2]]:
        ax.plot([0,end[0]],[0,end[1]],[0,end[2]], 'gray', alpha=0.3, lw=0.8)

    ax.text(0, 0,  1.35, '|0⟩', ha='center', fontsize=10, color='white')
    ax.text(0, 0, -1.35, '|1⟩', ha='center', fontsize=10, color='white')
    ax.text(1.45, 0, 0,  '|+⟩', ha='center', fontsize=8,  color='gray', alpha=0.7)

    bx, by, bz = bloch_vec
    ax.quiver(0, 0, 0, bx, by, bz, color=color, linewidth=2.5,
              arrow_length_ratio=0.15, zorder=5)
    ax.scatter([bx], [by], [bz], color=color, s=40, zorder=6)

    ax.set_title(title, fontsize=10, color='white', pad=8)
    ax.set_xlim([-1.5, 1.5]); ax.set_ylim([-1.5, 1.5]); ax.set_zlim([-1.5, 1.5])
    ax.set_axis_off()
    ax.set_facecolor('#111111')
```

```python
fig = plt.figure(figsize=(15, 5), facecolor='#111111')
fig.patch.set_facecolor('#111111')

# ── Passo 1: estado inicial ──────────────────────────────────────────────────
psi_0 = np.array([1+0j, 0+0j])          # |0⟩
ax1 = fig.add_subplot(131, projection='3d')
draw_bloch(ax1, state_to_bloch(*psi_0), "1. Estado inicial\n|ψ₀⟩ = |0⟩", color='dodgerblue')

# ── Passo 2: porta Hadamard ──────────────────────────────────────────────────
H = np.array([[1, 1], [1, -1]]) / np.sqrt(2)
psi_1 = H @ psi_0                        # H|0⟩ = |+⟩
ax2 = fig.add_subplot(132, projection='3d')
draw_bloch(ax2, state_to_bloch(*psi_1), "2. Após porta H\n|ψ₁⟩ = |+⟩", color='crimson')

# ── Passo 3: medição ─────────────────────────────────────────────────────────
np.random.seed(7)                        # fixar para reprodutibilidade
probs = [abs(psi_1[0])**2, abs(psi_1[1])**2]
resultado = np.random.choice([0, 1], p=probs)
psi_f = np.array([1+0j, 0+0j]) if resultado == 0 else np.array([0+0j, 1+0j])
ax3 = fig.add_subplot(133, projection='3d')
draw_bloch(ax3, state_to_bloch(*psi_f), f"3. Após medição\n→ |{resultado}⟩", color='gold')

plt.suptitle("Ciclo de vida de um qubit: preparação → operação → leitura",
             color='white', fontsize=12, y=1.0)
plt.tight_layout()
plt.show()

print(f"Amplitudes antes da medição:  α = {psi_1[0]:.4f},  β = {psi_1[1]:.4f}")
print(f"Probabilidades:               P(|0⟩) = {probs[0]:.0%},  P(|1⟩) = {probs[1]:.0%}")
print(f"Resultado da medição:         |{resultado}⟩  ← o vetor colapsou para o polo {'norte' if resultado == 0 else 'sul'}")
```
