## O que é

O comutador de dois operadores é [A,B] = AB - BA. Se [A,B] = 0, eles comutam e e^{A+B} = e^A·e^B exatamente. Se [A,B] ≠ 0, a decomposição de Trotter introduz erro proporcional a ||[A,B]||.

## Por que importa aqui

C-05 usa decomposição de Trotter para simular e^{-iHt}. O erro de Trotter não é arbitrário — ele é controlado pelo comutador dos termos de H. Esta digressão torna o bound de erro concreto.
