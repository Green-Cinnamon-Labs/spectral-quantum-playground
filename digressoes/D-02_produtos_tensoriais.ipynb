## O que é

O produto tensorial A⊗B de operadores em espaços H_A e H_B é um operador no espaço composto H_A⊗H_B. Para vetores: (a⊗b)ᵢⱼ = aᵢ·bⱼ.

## Por que importa aqui

Phase kickback acontece em um sistema de dois registradores — controle e alvo. Sem produto tensorial, não é possível descrever o estado conjunto nem a operação ctrl-U com precisão.
