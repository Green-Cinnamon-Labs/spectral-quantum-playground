## O que é

Um block-encoding de A é um unitário U tal que U[:n,:n] = A/α para algum fator de normalização α. Permite embutir matrizes não-unitárias em circuitos quânticos.

## Por que importa aqui

QSVT opera sobre valores singulares de matrizes arbitrárias — mas circuitos só podem implementar unitários. Block-encoding é o mecanismo que resolve essa tensão. C-07 e C-08 assumem que você entende o que o bloco representa.
