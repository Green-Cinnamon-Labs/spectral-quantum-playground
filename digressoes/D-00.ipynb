## Afirmação introdutória

Um `qubit` é qualquer sistema físico de dois níveis que obedece à mecânica quântica. Matematicamente, é um vetor normalizado em $\mathbb{C}^2$ — um espaço de Hilbert.

$$|\psi\rangle = \alpha|0\rangle + \beta|1\rangle, \quad \alpha, \beta \in \mathbb{C}, \quad |\alpha|^2 + |\beta|^2 = 1$$

<div style="max-width: 720px; margin: 0 auto; text-align: center; font-size: 0.9em; color: #aaa; line-height: 1.5;">
Um vetor de estado que é uma composição de dois outros vetores multiplicados por coeficientes que são limitados por duas regras.
</div>

Nesse notebook, vamos explorar as noções para essa expressão e direcionar você aos conceitos que precisa entender.

## 1. Explicação da expressão

- $|\psi\rangle$ é um vetor de estado em potencial;
- $|0\rangle$ e $|1\rangle$ são dois vetores definidos — não em potencial — que serão a base de $|\psi\rangle$; 
- $\alpha, \beta$ são coeficientes que dizem "quanto" de cada vetor acima existe no vetor de estado;
- $\in \mathbb{C}$ indica que eles *pertencem* aos números complexos, são dois números complexos;
- $|\alpha|^2 + |\beta|^2 = 1$ indica uma limitação para $\alpha, \beta$. Eles tem que ser tais que o quadrado do módulo de sua soma dê 1, sempre. Isso tem forte conotação com probabilidade, que sempre soma 100%

Dizer “um sistema físico” é só uma forma mais formal de dizer “uma coisa real do mundo”, algo palpável, que existe de verdade. E no caso, este sistema seria especial porque manifesta-se de acordo com as regras, com as leis, da mecânica quântica. 

Exemplos físicos: 
- spin de elétron ($\uparrow$/$\downarrow$)
- polarização de fóton (horizontal/vertical)
- dois níveis de energia de átomo. 

Todos são objetos físicos e que se manifestam através de uma mesma matemática, a mecânica quântica. Com efeito, você verá até o final desse notebook que existem diversas implementações físicas desse tal qubit.

O símbolo $|\psi\rangle$ expressa um objeto matemático avançado e especial chamado `vetor de estado`, ele é um vetor em um espaço complexo específico conhecido como `Espaço de Hilbert`.

## 2. Qual é a relação desse vetor com o qubit?

Este vetor e sua aplicação vieram muito antes do qubit — mas isso eu explico aqui <a href="../matematica/M-00.ipynb" style="font-size: 0.82em; color: #6b7280; text-decoration: none; border-bottom: 1px dashed #6b7280; white-space: nowrap;">↳ Vetores, Espaços e Espaço de Hilbert</a>. Uma pergunta mais efetiva seria: **o que eu ganho ao escolher** um vetor de estado para como objeto matemático para representar um qubit? Eu vou explicar o que se ganha com essa escolha, mas quero dizer que o ganho não é gratuito: você terá que fazer o qubit, de fato, comportar-se como um vetor de estado, e isso implica fazê-lo ser um sistema quântico — o que é bem difícil.

Um sistema físico que é regido pela mecânica quântica é bem pouco intuitivo. Trata-se de uma escala de grandezas tão ínfima que algumas leis clássicas de movimento deixam de descrever o que está acontecendo. Se você pergunta "qual é a trajetória dessa partícula?" a resposta deixa de ser um caminho bem definido no espaço ao longo do tempo e passa a ser uma descrição probabilística das possíveis posições e comportamentos do sistema. 

Com efeito, um dos princípios da Mecânica Quântica diz que não é possível determinar simultaneamente, com precisão arbitrária, a posição e o momento linear de uma partícula. É o famoso `Princípio da Incerteza de Heisenberg`, expresso por:

$$
\Delta x \, \Delta p \geq \frac{\hbar}{2}
$$

<div style="max-width: 720px; margin: 0 auto; text-align: center; font-size: 0.9em; color: #aaa; line-height: 1.5;">
  O produto da <i>Posição</i> pelo <i>Momento</i> nunca pode ser menor que um limite. Se tentamos reduzir demais um deles, o outro precisa crescer para manter esse limite.
</div>

- $\Delta x $ expressa uma variação de posição de algo, é uma taxa na qual a posição disso está mudando, no instante avaliado;
- $\Delta p $ expressa uma variação do momento de algo, é uma taxa na qual o momento disso está mudando, no instante avaliado;
- $\Delta x \, \Delta p \geq $ expressa que o produto dessas taxas é igual ou maior do que algo;
- $\frac{\hbar}{2} $ `constante de Planck` divida por 2 ;

Quando digo que não é possível, quero dizer que a própria natureza não permite isso. Não é como se eu estivesse dizendo "nossos equipamentos não conseguem"; estou dizendo que a realidade deliberadamente não fornece uma resolução. Não importa quão poderoso seja o instrumento: nem mesmo um observador perfeito — um anjo, ou até Deus — seria capaz de determinar isso com precisão arbitrária. É como se a coisa em questão fosse tão pequena que a realidade não consegue resolvê-la e, sendo assim, não permite que você a conheça exatamente, preferindo escondê-la de você.

Aí é que está a malandragem, o pulo do gato: ela esconde isso de você colocando a coisa em superposição. Ela faz com que aquela coisa seja todas as possibilidades ao mesmo tempo, faz aquilo ser potência e não fato. É por isso que fazer o qubit ser isso — seja lá o que for — é interessante, porque você consegue avaliar várias opções em superposição; é mais rápido. Pense assim: você precisa descobrir quem, na sala, se chama *Matheus*. Você pode ir um por um perguntando, ou pode gritar — o que equivale a perguntar a todos ao mesmo tempo. Isso é bem interessante para aplicação em computação porque assim encontramos os resultados mais rápido.

```python
import numpy as np
import matplotlib.pyplot as plt

# ── Problema: busca não-estruturada ──────────────────────────────────────────
# Dado um espaço de N candidatas, encontrar a única correta.
#
# Clássico (i9, 24 núcleos): testa uma por uma em paralelo → O(N / 24)
# Quântico (Grover):         √N iterações                  → O(√N)

cores_i9 = 24
qubits_range = np.arange(4, 51)          # de 4 a 50 qubits
N = 2.0 ** qubits_range                  # espaço de busca: 2^n candidatas

ops_classico = N / cores_i9              # operações clássicas (paralelo)
ops_grover   = np.sqrt(N)                # iterações de Grover

speedup = ops_classico / ops_grover      # quantas vezes mais rápido

# ── Gráfico ──────────────────────────────────────────────────────────────────
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 5), facecolor='#111')
fig.patch.set_facecolor('#111')

for ax in (ax1, ax2):
    ax.set_facecolor('#111')
    for spine in ax.spines.values():
        spine.set_edgecolor('#374151')
    ax.tick_params(colors='#9ca3af')
    ax.grid(True, color='#1f2937', lw=0.5)

# Painel 1: operações necessárias (escala log)
ax1.semilogy(qubits_range, ops_classico, color='tomato',    lw=2.5, label=f'CPU i9 ({cores_i9} núcleos) — O(N/{cores_i9})')
ax1.semilogy(qubits_range, ops_grover,   color='steelblue', lw=2.5, label='QPU Grover — O(√N)')
ax1.set_xlabel('número de qubits (n)', color='#9ca3af', fontsize=10)
ax1.set_ylabel('operações necessárias', color='#9ca3af', fontsize=10)
ax1.set_title('Operações para encontrar 1 solução entre 2ⁿ candidatas', color='white', fontsize=11)
ax1.legend(fontsize=9, labelcolor='#d1d5db', facecolor='#1f2937', edgecolor='#374151')

# Anotações em pontos específicos
for n in [10, 20, 30, 40, 50]:
    idx = n - 4
    ax1.annotate(f'n={n}', xy=(n, ops_grover[idx]),
                 xytext=(n + 1, ops_grover[idx] * 5),
                 color='steelblue', fontsize=7.5,
                 arrowprops=dict(arrowstyle='->', color='steelblue', lw=0.8))

# Painel 2: speedup
ax2.semilogy(qubits_range, speedup, color='#f59e0b', lw=2.5)
ax2.set_xlabel('número de qubits (n)', color='#9ca3af', fontsize=10)
ax2.set_ylabel('speedup (vezes mais rápido)', color='#9ca3af', fontsize=10)
ax2.set_title('Vantagem quântica: QPU vs CPU i9', color='white', fontsize=11)
ax2.fill_between(qubits_range, 1, speedup, color='#f59e0b', alpha=0.1)

# Tabela legível
cell_text = [
    ['10',  '1.024',           '43',              '32',          '1x'],
    ['20',  '~1 milhão',       '43.691',          '1.024',       '43x'],
    ['30',  '~1 bilhão',       '44,7 milhões',    '32.768',      '1.365x'],
    ['40',  '~1 trilhão',      '45,8 bilhões',    '1 milhão',    '43.691x'],
    ['50',  '~1 quatrilhão',   '46,9 trilhões',   '33,5 milhões','1,4 milhão x'],
]
col_labels = ['qubits', 'candidatas', 'CPU i9', 'Grover', 'speedup']

table = ax2.table(cellText=cell_text, colLabels=col_labels,
                  cellLoc='center', loc='upper left',
                  bbox=[0.03, 0.45, 0.94, 0.52])
table.auto_set_font_size(False)
table.set_fontsize(8)
for cell in table.get_celld().values():
    cell.set_facecolor('#1f2937')
    cell.set_edgecolor('#374151')
    cell.set_text_props(color='#d1d5db')

plt.suptitle(
    'Busca não-estruturada:  clássico O(N)  vs  Grover O(√N)',
    color='#9ca3af', fontsize=12, y=1.02
)
plt.tight_layout()
plt.show()
```

O gráfico da esquerda mostra a quantidade de operações que cada abordagem precisa para encontrar 1 resposta certa entre $2^{n}$ candidatas. As duas linhas começam próximas, mas se separam rápido:
 - a vermelha (CPU i9, 24 núcleos) cresce exponencialmente com $n$;
- enquanto a azul (Grover) cresce com a raiz quadrada. 

Com 50 qubits, o espaço de busca tem mais de um quatrilhão de candidatas — o i9 precisa de ~47 trilhões de operações, Grover precisa de ~33 milhões. O gráfico da direita traduz essa diferença em speedup: a vantagem quântica não é fixa, ela cresce exponencialmente conforme o problema cresce. 

Dito de outro modo: adicionar 1 qubit dobra o espaço de busca para o clássico, mas só multiplica o trabalho do quântico por √2. Tudo isso porque você consegue avaliar as coisas simultaneamente. É essa assimetria que torna a computação quântica interessante — não para problemas pequenos, mas para problemas onde o espaço de candidatas escala exponencialmente.

## 3. Os limites do qubit

Em 1981, muitos pesquisadores estavam interessados em um método que permitisse simular como moléculas e processos químicos evoluem no tempo. A questão é que essa evolução é justamente determinada pela mecânica quântica. Descrever essa evolução num computador clássico fica rapidamente inviável do ponto de vista de memória e poder computacional, levaria muito tempo.

Nesse contexto, surgiu a famosa proposta de Richard Feynman: usar um sistema quântico para simular outro sistema quântico. A ideia era controlá-lo, deixá-lo evoluir segundo a dinâmica do sistema de interesse e, depois, medi-lo. Aqui, o foco ainda era a simulação de sistemas físicos quânticos, não ainda a formulação de um modelo geral de computação quântica.

A partir daí, começou a busca por um sistema físico simples, controlável e legível, que também obedecesse à mecânica quântica. Em essência, havia apenas três exigências:

1. Ser preparável em um estado inicial conhecido;
2. Ser manipulável de forma previsível ao longo do tempo;
3. Ser medível no fim, sem ambiguidade;

Em 1985, David Deutsch formaliza o computador quântico universal. Ele define o que seria uma `máquina de Turing quântica` e mostra que ela pode simular qualquer sistema quântico eficientemente. **Aqui nasce a computação quântica** como modelo de computação, não apenas simulação. No mesmo ano, Asher Peres publica um artigo sobre lógica reversível e computadores quânticos, onde já aparece o primeiro `código de correção de erros` de bit (o código de 3 qubits).

Em 1992, Deutsch e Jozsa formularam o problema que ficaria conhecido como `algoritmo de Deutsch–Jozsa`, o primeiro exemplo célebre de vantagem quântica demonstrável em um problema de consulta. O problema é artificial, mas a separação entre clássico e quântico é nítida: na formulação mais conhecida hoje, refinada depois por Cleve, Ekert, Macchiavello e Mosca, o algoritmo quântico resolve o problema com uma única consulta ao `oráculo`, enquanto o caso clássico exige várias. No mesmo ano, Bernstein e Vazirani deram rigor matemático ao modelo de máquina de Turing quântica; a partir daí, Peter Shor começou a prestar atenção.


Em 1994, Dan Simon resolve o problema de `periodicidade num hipercubo` usando `transformada de Fourier quântica`. A separação é exponencial: classicamente o problema é intratável, quanticamente é eficiente. O paper é rejeitado na conferência STOC — Peter Shor está no comitê que o rejeitou.

Em 1994, Shor adapta a ideia de Simon. Ele percebe que `fatoração` e `logaritmo discreto` também são problemas de periodicidade — só que num grupo cíclico, não num hipercubo. Adapta a transformada de Fourier quântica para esse domínio. Resolve o logaritmo discreto primeiro, depois a fatoração. **O resultado quebra RSA em tempo polinomial** num computador quântico. A `criptografia de chave pública`, como era praticada, ficou teoricamente vulnerável.

Em 1995, A objeção de Landauer. O `teorema de não-clonagem` parece impedir qualquer correção de erros quânticos. Sem correção, o algoritmo de Shor exigiria portas com precisão de uma parte em um bilhão — fisicamente impossível. Shor responde com o código de 9 qubits: concatena um código que corrige erros de bit com um que corrige erros de fase. Refuta a objeção. Calderbank, Shor e Steane generalizam para a família de `códigos CSS`.

Em 1996, Grover publica o `algoritmo de busca`. Speedup quadrático para busca não-estruturada. Menos dramático que Shor, mas aplicável a uma classe muito mais ampla de problemas.

## 4. Enquanto isso: a arquitetura do qubit

Em 1985, Grupo de John Clarke em Berkeley (com Michel Devoret e John Martinis) faz a primeira espectroscopia dos níveis de energia quantizados de uma `junção de Josephson`. Enviam micro-ondas para excitar transições entre níveis discretos e detectam o resultado via `tunelamento quântico`. É a primeira evidência experimental direta de que um circuito macroscópico tem espectro quantizado — premissa necessária para qualquer `qubit supercondutor`.

Em 1995, Cirac e Zoller propõem a primeira arquitetura concreta: `íons aprisionados` manipulados por laser. Dois níveis eletrônicos do íon são os estados do qubit. É a primeira proposta que mostra como implementar `portas universais` num sistema físico real.

Em 1996, as primeiras demonstrações experimentais com ressonância magnética nuclear (NMR). Limitada em escalabilidade, mas permitiu rodar algoritmos pequenos (2-7 qubits) e validar a teoria.

Em 1997, Kitaev propõe `computação topológica`. A ideia é codificar o qubit não num sistema local, mas na topologia global de `quasipartículas (anyons)`. Erros locais não afetam a informação. Altamente resistente a ruído, mas ninguém conseguiu construir um até hoje.

Em 1999, três marcos simultâneos em qubits supercondutores. Nakamura et al. demonstram `oscilações de Rabi` coerentes numa caixa de `pares de Cooper` (charge qubit) — a primeira evidência direta, no domínio do tempo, de coerência quântica num circuito eletrônico. As oscilações duraram apenas alguns nanossegundos, mas mostraram que era possível. No mesmo ano, o grupo de Hans Mooij em Delft propõe o `flux qubit`: um loop supercondutor com três junções de Josephson, onde o qubit é a superposição de corrente circulando nos dois sentidos. Martinis, por sua vez, desenvolve o `phase qubit`, explorando os dois níveis mais baixos do `potencial de washboard`.

Em 2002, o grupo Saclay (Devoret, Esteve, Vion) demonstra o `quantronium`. Um qubit híbrido carga-fase com um "ponto doce" onde a frequência de transição é insensível a ruído nos parâmetros de controle. Mostram as primeiras `franjas de interferência de Ramsey` em qubits supercondutores — o teste definitivo de coerência. `Tempo de coerência`: microsegundos, um fator ~100 acima do experimento de Nakamura.

Em 2007, Koch et al. propõem o `transmon`. É, no fundo, uma caixa de pares de Cooper com um capacitor grande na forma de uma antena dipolar de ~1 mm. Isso o torna exponencialmente insensível a ruído de carga, ao custo de uma redução modesta na `anarmonicidade`. O momento de dipolo é ~100.000 vezes maior que o de um átomo natural, o que garante acoplamento forte com fótons de micro-ondas. É a arquitetura dominante hoje (IBM, Google). Nos últimos 20 anos, os tempos de coerência melhoraram cerca de seis ordens de magnitude.
