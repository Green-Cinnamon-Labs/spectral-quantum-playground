# D-02 — Criptografia e Logaritmos

- Multiplicar dois primos é trivial: $37 \times 53 = 1961$. 
- Fatorar o produto — dado $1961$, encontrar $37$ e $53$ — é difícil. 

Tão difícil que toda a criptografia de chave pública moderna (RSA) aposta nisso: **qualquer pessoa pode gerar o produto, mas ninguém consegue desfazer a multiplicação em tempo razoável**. O melhor algoritmo clássico conhecido, o *number field sieve*, leva um tempo que é dado por essa fórmula:


$$\exp\!\Big(c\,(\log N)^{1/3}(\log\log N)^{2/3}\Big)$$

- $N$ — o inteiro a fatorar.
- $\log N$ — uma medida do tamanho da entrada: número de bits (se o log for na base 2) ou, a menos de constante, número de dígitos.
- $c$ — constante fixa do algoritmo ($c=(64/9)^{1/3}\approx 1.923$ para o *general number field sieve*).
- $(\log N)^{1/3}$ — parte principal do expoente; é isso que impede o crescimento de virar exponencial puro no tamanho da entrada.
- $(\log\log N)^{2/3}$ — fator de correção que cresce extremamente devagar.
- $\exp(\dots)$ — crescimento **subexponencial**: mais rápido que qualquer polinômio no tamanho da entrada, mas mais lento que $N^k$ para qualquer constante fixa $k$.

Para RSA-2048, o expoente fica em $\sim 110$, e $e^{110}$ é astronômico — intratável.

Em 1994, Peter Shor mostrou que um computador quântico pode fatorar inteiros em tempo polinomial, isto é, um tempo dado por essa fórmula:

$$O((\log N)^2 \log\log N \cdot \log\log\log N)$$

- $(\log N)^2$ — o custo dominante; vem da exponenciação modular ($x^a \bmod N$), que faz $O(\log N)$ multiplicações de números com $O(\log N)$ dígitos.
- $\log\log N$ — vem do algoritmo de multiplicação rápida (Schönhage–Strassen) usado dentro de cada multiplicação modular.
- $\log\log\log N$ — fator de correção do próprio Schönhage–Strassen; na prática é $\leq 4$ para qualquer $N$ concebível.

O ponto central: todos os termos são **polinômios e logaritmos** de $\log N$. Nada é exponencial. Para RSA-2048, $(\log N)^2 \approx 2048^2 \approx 4 \times 10^6$ — um número que um computador processa sem dificuldade.
 
Esse resultado sacudiu a área inteira. Não porque alguém quisesse quebrar RSA amanhã, mas porque demonstrou que a forma como você codifica informação em qubits pode transformar um problema intratável em um problema eficiente.

A pergunta que nos interessa aqui é: **como Shor codificou o problema de fatoração usando qubits?** A resposta revela a essência da codificação quântica.
