## O que é

Bom, primeiro vou fizer o que é formalmente e depois vou traduzir.

Um `qubit` é qualquer sistema físico de dois níveis que obedece à mecânica quântica. Matematicamente, é um vetor normalizado em $\mathbb{C}^2$:

$$|\psi\rangle = \alpha|0\rangle + \beta|1\rangle, \quad \alpha, \beta \in \mathbb{C}, \quad |\alpha|^2 + |\beta|^2 = 1$$

Exemplos físicos: spin de elétron ($\uparrow$/$\downarrow$), polarização de fóton (horizontal/vertical), dois níveis de energia de átomo. O que os une é a mesma estrutura matemática.

## Por que importa aqui

C-00 usa o qubit antes de defini-lo. Esta digressão fecha essa lacuna: o qubit não é "um bit que pode ser 0 e 1 ao mesmo tempo" — é um vetor em $\mathbb{C}^2$ cuja riqueza vem da estrutura complexa, não da superposição em si.

A representação geométrica canônica é a **esfera de Bloch**: todo estado puro de 1 qubit corresponde a um ponto na superfície de uma esfera unitária em $\mathbb{R}^3$, parametrizado por dois ângulos reais $\theta$ e $\phi$:

$$|\psi\rangle = \cos\frac{\theta}{2}|0\rangle + e^{i\phi}\sin\frac{\theta}{2}|1\rangle$$

Os polos são $|0\rangle$ (norte) e $|1\rangle$ (sul). Rotações na esfera de Bloch correspondem a portas quânticas.
