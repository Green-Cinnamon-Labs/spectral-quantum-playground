## O que é

A Born rule estabelece que medir um estado |ψ⟩ = Σ cₖ|k⟩ na base {|k⟩} retorna o resultado k com probabilidade |cₖ|². O estado colapsa para |k⟩ após a medição.

## Por que importa aqui

C-01 assume que você sabe o que acontece quando medimos. Esta digressão torna isso preciso: define o colapso, distingue medição projetiva de POVM, e mostra que o colapso é irreversível — o que justifica por que toda a computação útil acontece *antes* da medição.
